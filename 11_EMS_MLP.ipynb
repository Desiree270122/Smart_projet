```python
%run "./10_EMS_alpha_star.ipynb"
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt

def mae(y, yh): return float(np.mean(np.abs(y - yh)))
def rmse(y, yh): return float(np.sqrt(np.mean((y - yh) ** 2)))
def r2(y, yh):
    den = np.sum((y - np.mean(y)) ** 2)
    return float(1 - np.sum((y - yh) ** 2) / den) if not np.isclose(den, 0) else np.nan

print('Cible utilisee : MLP_SIMPLE_TARGET =', MLP_SIMPLE_TARGET)
print(optimized_datasets['train'].shape, optimized_datasets['validation'].shape, optimized_datasets['test'].shape)
```

construction X/y (seulement les lignes ou alpha_star est defini)

```python
 
def build_xy(data, cols, target_col):
    valid = data[target_col].notna() & data['alpha_ems_alpha_star_valid'].eq(1)
    X = data.loc[valid, cols].to_numpy(dtype=np.float32)
    y = data.loc[valid, target_col].to_numpy(dtype=np.float32)
    return X, y, valid

X_train, y_train, valid_train = build_xy(optimized_datasets['train'], MLP_SIMPLE_INPUT_COLS, MLP_SIMPLE_TARGET)
X_val, y_val, valid_val = build_xy(optimized_datasets['validation'], MLP_SIMPLE_INPUT_COLS, MLP_SIMPLE_TARGET)
X_test, y_test, valid_test = build_xy(optimized_datasets['test'], MLP_SIMPLE_INPUT_COLS, MLP_SIMPLE_TARGET)

print('Train :', X_train.shape, '| Val :', X_val.shape, '| Test :', X_test.shape)
```

normalisation des entrees (la cible alpha_star est deja dans [0,1], pas normalisee)

```python

x_mean = X_train.mean(axis=0)
x_std = X_train.std(axis=0)

if np.any(x_std < 1e-12):
    constant_cols = np.array(MLP_SIMPLE_INPUT_COLS)[x_std < 1e-12]
    raise ValueError(f"Variables constantes dans le train : {constant_cols.tolist()}")

def norm_x(X): return ((X - x_mean) / x_std).astype(np.float32)

Xn_train, Xn_val, Xn_test = norm_x(X_train), norm_x(X_val), norm_x(X_test)

MLP_SIMPLE_SCALER_FILE = MODELS_DIR / 'EMS_MLP_scalers.npz'
np.savez(MLP_SIMPLE_SCALER_FILE, x_mean=x_mean, x_std=x_std, input_cols=np.array(MLP_SIMPLE_INPUT_COLS))
print('x_mean :', x_mean)
print('x_std  :', x_std)
```

Dataset et DataLoader

modele (purement tabulaire, pas d'ontologie, pas de logique floue, pas de LSTM)

```python
class TabularDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X)
        self.y = torch.from_numpy(y)
    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]

generator = torch.Generator()
generator.manual_seed(RANDOM_SEED)

train_loader = DataLoader(TabularDataset(Xn_train, y_train), batch_size=MLP_SIMPLE_BATCH_SIZE,
                           shuffle=True, generator=generator)
val_loader = DataLoader(TabularDataset(Xn_val, y_val), batch_size=MLP_SIMPLE_BATCH_SIZE, shuffle=False)
test_loader = DataLoader(TabularDataset(Xn_test, y_test), batch_size=MLP_SIMPLE_BATCH_SIZE, shuffle=False)

print(len(train_loader.dataset), len(val_loader.dataset), len(test_loader.dataset))
```

```python
class MLPSimple(nn.Module):
    def __init__(self, n_in, hidden1, hidden2, dropout):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n_in, hidden1), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(hidden1, hidden2), nn.ReLU(),
            nn.Linear(hidden2, 1),
        )
    def forward(self, x):
        return torch.sigmoid(self.net(x)).squeeze(-1)

torch.manual_seed(RANDOM_SEED)
model = MLPSimple(len(MLP_SIMPLE_INPUT_COLS), MLP_SIMPLE_HIDDEN_1, MLP_SIMPLE_HIDDEN_2, MLP_SIMPLE_DROPOUT).to(DEVICE)
print(model)
```

 entrainement

```python

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=MLP_SIMPLE_LEARNING_RATE)

best_val = float('inf')
best_epoch = -1
patience_count = 0
history = {'train': [], 'val': []}

for epoch in range(MLP_SIMPLE_EPOCHS):
    model.train()
    train_loss = 0.0
    for xb, yb in train_loader:
        xb, yb = xb.to(DEVICE), yb.to(DEVICE)
        optimizer.zero_grad()
        pred = model(xb)
        loss = criterion(pred, yb)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        train_loss += loss.item() * xb.size(0)
    train_loss /= len(train_loader.dataset)

    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            val_loss += criterion(model(xb), yb).item() * xb.size(0)
    val_loss /= len(val_loader.dataset)

    history['train'].append(train_loss)
    history['val'].append(val_loss)

    if not np.isfinite(train_loss) or not np.isfinite(val_loss):
        raise ValueError("Loss non finie detectee.")

    if val_loss < best_val - 1e-6:
        best_val, best_epoch, patience_count = val_loss, epoch, 0
        torch.save(model.state_dict(), MLP_SIMPLE_CHECKPOINT)
    else:
        patience_count += 1

    if epoch == 0 or (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch+1:3d}/{MLP_SIMPLE_EPOCHS} train={train_loss:.6f} val={val_loss:.6f} patience={patience_count}")

    if patience_count >= MLP_SIMPLE_PATIENCE:
        print(f"Arret anticipe a l'epoque {epoch+1}.")
        break

try:
    state_dict = torch.load(MLP_SIMPLE_CHECKPOINT, map_location=DEVICE, weights_only=True)
except TypeError:
    state_dict = torch.load(MLP_SIMPLE_CHECKPOINT, map_location=DEVICE)
model.load_state_dict(state_dict)
print('Meilleure epoque :', best_epoch + 1)
print('Meilleure val loss :', best_val)
```

evaluation de regression (vs alpha_star)

```python

def predict(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for xb, _ in loader:
            preds.append(model(xb.to(DEVICE)).cpu().numpy())
    return np.concatenate(preds, axis=0)

pred_test = predict(model, test_loader)
print(f'MAE={mae(y_test, pred_test):.4f}  RMSE={rmse(y_test, pred_test):.4f}  R2={r2(y_test, pred_test):.4f}')
```
