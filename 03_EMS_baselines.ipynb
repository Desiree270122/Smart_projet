```python


import numpy as np
import pandas as pd

datasets = {
    "train": pd.read_csv(PROCESSED_DIR / "train.csv"),
    "validation": pd.read_csv(PROCESSED_DIR / "validation.csv"),
    "test": pd.read_csv(PROCESSED_DIR / "test.csv"),
} 

for name, data in datasets.items():
    print(name, data.shape)
```

```python
required_cols = [
    "time",
    "hasPower",
    "P_PB",
    "SOC_EB",
    "SOC_PB",
]

for name, data in datasets.items():
    missing = [
        col for col in required_cols
        if col not in data.columns
    ]

    if missing:
        raise ValueError(
            f"{name} : colonnes absentes {missing}"
        )

print("Colonnes requises disponibles.")
```

```python
def mae(y_true, y_pred):
    return float(
        np.mean(np.abs(y_true - y_pred))
    )


def rmse(y_true, y_pred):
    return float(
        np.sqrt(np.mean((y_true - y_pred) ** 2))
    )


def r2_score_np(y_true, y_pred):
    denominator = np.sum(
        (y_true - np.mean(y_true)) ** 2
    )

    if np.isclose(denominator, 0.0):
        return np.nan

    numerator = np.sum(
        (y_true - y_pred) ** 2
    )

    return float(
        1.0 - numerator / denominator
    )
```

```python
def persistence_eval(data, window, horizon):
    p_dem = data["hasPower"].to_numpy()
    soc_eb = data["SOC_EB"].to_numpy()
    soc_pb = data["SOC_PB"].to_numpy()

    anchors = np.arange(
        window - 1,
        len(data) - horizon
    )

    targets = anchors + horizon

    return {
        "Pdem": (
            p_dem[targets],
            p_dem[anchors],
        ),
        "delta_SOC_EB": (
            soc_eb[targets] - soc_eb[anchors],
            np.zeros(len(anchors)),
        ),
        "delta_SOC_PB": (
            soc_pb[targets] - soc_pb[anchors],
            np.zeros(len(anchors)),
        ),
    }
```

```python
metrics_rows = []

for split_name, data in datasets.items():
    results = persistence_eval(
        data,
        window=WINDOW,
        horizon=HORIZON,
    )

    print(f"\n--- {split_name} ---")

    for target_name, values in results.items():
        y_true, y_pred = values

        row = {
            "split": split_name,
            "target": target_name,
            "n": len(y_true),
            "mae": mae(y_true, y_pred),
            "rmse": rmse(y_true, y_pred),
            "r2": r2_score_np(y_true, y_pred),
        }

        metrics_rows.append(row)

        print(
            f"{target_name:15s} "
            f"n={row['n']} "
            f"MAE={row['mae']:.6f} "
            f"RMSE={row['rmse']:.6f} "
            f"R2={row['r2']:.4f}"
        )

persistence_metrics = pd.DataFrame(
    metrics_rows
)

display(persistence_metrics)
```

```python
def add_historical_alpha(data, eps):
    result = data.copy()

    p_dem = result["hasPower"].to_numpy()
    p_pb = result["P_PB"].to_numpy()

    valid = np.abs(p_dem) > eps

    alpha = np.full(
        len(result),
        np.nan,
        dtype=float,
    )

    alpha[valid] = (
        p_pb[valid] / p_dem[valid]
    )

    alpha[valid] = np.clip(
        alpha[valid],
        ALPHA_MIN,
        ALPHA_MAX,
    )

    result["alpha_ems_eb_priority"] = alpha
    result["alpha_ems_eb_priority_valid"] = (
        valid.astype(np.int8)
    )

    return result
```
