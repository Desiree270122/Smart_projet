```python
# === 15_evaluation_comparative.ipynb ===
# Cellule 1 : chargement
%run "./13_EMS_safety_filter.ipynb"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def mae(y, yh): return float(np.mean(np.abs(y - yh)))
def rmse(y, yh): return float(np.sqrt(np.mean((y - yh) ** 2)))
def r2(y, yh):
    den = np.sum((y - np.mean(y)) ** 2)
    return float(1 - np.sum((y - yh) ** 2) / den) if not np.isclose(den, 0) else np.nan

print('test_sub :', test_sub.shape)
```

```python
# Cellule 2 : recuperer alpha_mlp_simple et alpha_gnn alignes sur test_sub (par le temps)
mlp_simple_file = PREDICTIONS_DIR / 'predictions_mlp_simple.csv'
gnn_file = PREDICTIONS_DIR / 'predictions_gnn_simple.csv'

base = test_sub[[TIME_COL, 'SOC_EB', 'SOC_PB', 'hasPower', 'alpha_ems_alpha_star', 'alpha_ems_eb_priority', 'alpha_ems_fuzzy_logic',
                  'J_star', 'J_fuzzy', 'J_priority', 'fuzzy_feasible', 'priority_feasible']].copy()
base['alpha_ems_mlp_neurosymbolic'] = alpha_mlp_ns_test
base['mlp_ns_feasible'] = mlp_ns_feasible
base['mlp_ns_cost'] = mlp_ns_cost

if mlp_simple_file.exists():
    mlp_simple_pred = pd.read_csv(mlp_simple_file)[[TIME_COL, 'alpha_ems_mlp']]
    base = base.merge(mlp_simple_pred, on=TIME_COL, how='left')
else:
    print('Fichier absent :', mlp_simple_file)
    base['alpha_ems_mlp'] = np.nan

if gnn_file.exists():
    gnn_pred = pd.read_csv(gnn_file)[[TIME_COL, 'alpha_ems_gnn']]
    base = base.merge(gnn_pred, on=TIME_COL, how='left')
else:
    print('Fichier absent :', gnn_file)
    base['alpha_ems_gnn'] = np.nan

print(base.shape)
print('lignes avec alpha_ems_mlp:', base['alpha_ems_mlp'].notna().sum())
print('lignes avec alpha_ems_gnn      :', base['alpha_ems_gnn'].notna().sum())
```

```python
# Cellule 3 : evaluer mlp_simple et gnn avec le meme cout (sequentiel, comme les autres)
def evaluate_sequence(alpha_series, p_series, soceb_series, socpb_series):
    prev = None
    feas = np.zeros(len(alpha_series), dtype=np.int8)
    cost = np.full(len(alpha_series), np.nan)
    for i in range(len(alpha_series)):
        a = alpha_series[i]
        if not np.isfinite(a):
            continue
        res = evaluate_alpha_row(a, p_series[i], soceb_series[i], socpb_series[i], prev)
        feas[i] = res['feasible']
        cost[i] = res['cost']
        prev = a
    return feas, cost

p_arr = base['hasPower'].to_numpy()
se_arr = base['SOC_EB'].to_numpy()
sp_arr = base['SOC_PB'].to_numpy()

mlp_simple_feas, mlp_simple_cost = evaluate_sequence(base['alpha_ems_mlp'].to_numpy(), p_arr, se_arr, sp_arr)
gnn_feas, gnn_cost = evaluate_sequence(base['alpha_ems_gnn'].to_numpy(), p_arr, se_arr, sp_arr)

base['mlp_simple_feasible'] = mlp_simple_feas
base['mlp_simple_cost'] = mlp_simple_cost
base['gnn_feasible'] = gnn_feas
base['gnn_cost'] = gnn_cost
```

```python
# === Balayage de MAX_DELTA_ALPHA -- reentrainement complet a chaque valeur, selection sur validation ===
MAX_DELTA_CANDIDATES = [0.20, 0.30, 0.40, 0.50, 0.75, 1.00]

def train_mlp_ns(max_delta, seed=RANDOM_SEED):
    torch.manual_seed(seed)
    m = MLPNeuroSymbolique(len(MLP_NS_INPUT_COLS), MLP_NS_HIDDEN_1, MLP_NS_HIDDEN_2, MLP_NS_DROPOUT, max_delta).to(DEVICE)
    opt = torch.optim.Adam(m.parameters(), lr=MLP_NS_LEARNING_RATE)
    best_val, best_state, patience_count = float('inf'), None, 0

    for epoch in range(MLP_NS_EPOCHS):
        m.train()
        for xb, fz, pd_, se, sp, ea, yb in train_loader:
            xb, fz, pd_, se, sp, ea, yb = [t.to(DEVICE) for t in (xb, fz, pd_, se, sp, ea, yb)]
            opt.zero_grad()
            alpha_pred, _ = m(xb, fz)
            alpha_prev = torch.cat([yb[:1], yb[:-1]])
            loss, _ = composite_loss(alpha_pred, yb, pd_, se, sp, ea, alpha_prev)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(m.parameters(), max_norm=1.0)
            opt.step()

        m.eval()
        val_loss = 0.0
        with torch.no_grad():
            for xb, fz, pd_, se, sp, ea, yb in val_loader:
                xb, fz, pd_, se, sp, ea, yb = [t.to(DEVICE) for t in (xb, fz, pd_, se, sp, ea, yb)]
                alpha_pred, _ = m(xb, fz)
                alpha_prev = torch.cat([yb[:1], yb[:-1]])
                loss, _ = composite_loss(alpha_pred, yb, pd_, se, sp, ea, alpha_prev)
                val_loss += loss.item() * xb.size(0)
        val_loss /= len(val_ds)

        if val_loss < best_val - 1e-6:
            best_val, best_state, patience_count = val_loss, {k: v.clone() for k, v in m.state_dict().items()}, 0
        else:
            patience_count += 1
        if patience_count >= MLP_NS_PATIENCE:
            break

    m.load_state_dict(best_state)
    m.eval()
    return m, best_val


def evaluate_max_delta(max_delta, model, X, alpha_fuzzy_arr, y_true, p_dem_arr, soc_eb_arr, soc_pb_arr, j_star_arr):
    with torch.no_grad():
        alpha_pred, _ = model(torch.from_numpy(X).to(DEVICE), torch.from_numpy(alpha_fuzzy_arr).to(DEVICE))
    alpha_pred = alpha_pred.cpu().numpy()

    lower = np.clip(alpha_fuzzy_arr - max_delta, 0.0, 1.0)
    upper = np.clip(alpha_fuzzy_arr + max_delta, 0.0, 1.0)
    reachable = (y_true >= lower - 1e-7) & (y_true <= upper + 1e-7)

    prev = None
    feas = np.zeros(len(alpha_pred), dtype=np.int8)
    cost = np.full(len(alpha_pred), np.nan)
    for i in range(len(alpha_pred)):
        res = evaluate_alpha_row(alpha_pred[i], p_dem_arr[i], soc_eb_arr[i], soc_pb_arr[i], prev)
        feas[i] = res['feasible']; cost[i] = res['cost']
        prev = alpha_pred[i]

    return {
        'max_delta': max_delta,
        'reachable_pct': 100.0 * reachable.mean(),
        'mae': mae(y_true, alpha_pred), 'rmse': rmse(y_true, alpha_pred), 'r2': r2(y_true, alpha_pred),
        'mean_cost': float(np.nanmean(cost)),
        'regret_vs_alpha_star': float(np.nanmean(cost) - np.nanmean(j_star_arr)),
        'feasibility_pct_avant_filtre': 100 * feas.mean(),
        'taux_correction_filtre_estime': 100 * (1 - feas.mean()),
        'sat_low_pct': float((alpha_pred <= 0.01).mean() * 100),
        'sat_high_pct': float((alpha_pred >= 0.99).mean() * 100),
    }


j_star_val = merged_datasets['validation'].loc[merged_datasets['validation']['Pdem_pred_ns'].notna() &
                                                merged_datasets['validation']['alpha_ems_alpha_star_valid'].eq(1), 'J_star'].to_numpy()



sweep_results = []
trained_models = {}
for md in MAX_DELTA_CANDIDATES:
    print('Entrainement avec MAX_DELTA_ALPHA =', md)
    m, best_val_loss = train_mlp_ns(md)
    trained_models[md] = m
    res = evaluate_max_delta(md, m, Xn_val, fuzzy_val, y_val, p_val, soceb_val, socpb_val, j_star_val)
    res['best_val_loss'] = best_val_loss
    sweep_results.append(res)
    print('  ', {k: round(v, 4) if isinstance(v, float) else v for k, v in res.items()})

sweep_df = pd.DataFrame(sweep_results)
display(sweep_df)
sweep_df.to_csv(TABLES_DIR / 'mlp_ns_max_delta_sweep.csv', index=False)
```
