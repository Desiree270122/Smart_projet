```python
# Cellule 2 — Définition des groupes d'entrées


PHYSICAL_COLS = [
    "hasPower",
    "SOC_EB",
    "SOC_PB",
    "speed",
    "hasAcceleration",
]

VARIABLE_SYMBOLIC_COLS = [
    "high_power_demand",
    "regenerative_braking",
    "zero_power_demand",
    "converter_risk",
]

LSTM_PREDICTION_COLS = [
    "Pdem_pred_ns",
    "delta_soc_eb_pred_ns",
    "delta_soc_pb_pred_ns",
]

FUZZY_COLS = [
    "alpha_ems_fuzzy_logic",
]

ABLATION_VARIANTS = {
    "MLP": (
        PHYSICAL_COLS
    ),

    "MLP_symbolique": (
        PHYSICAL_COLS
        + VARIABLE_SYMBOLIC_COLS
    ),

    "MLP_LSTM": (
        PHYSICAL_COLS
        + LSTM_PREDICTION_COLS
    ),

    "MLP_fuzzy": (
        PHYSICAL_COLS
        + FUZZY_COLS
    ),

    "MLP_neurosymbolique": (
        PHYSICAL_COLS
        + VARIABLE_SYMBOLIC_COLS
        + LSTM_PREDICTION_COLS
        + FUZZY_COLS
    ),
}

ABLATION_TARGET = "alpha_ems_alpha_star"

FORBIDDEN_INPUTS = {
    "alpha_ems_alpha_star",
    "alpha_ems_eb_priority",
    "P_EB",
    "P_PB",
    "P_conv",
    "I_EB",
    "I_PB",
    "P_EB_ems_alpha_star",
    "P_PB_ems_alpha_star",
    "P_conv_star",
    "J_star",
    "J_fuzzy",
    "J_priority",
}

for variant_name, columns in ABLATION_VARIANTS.items():
    leakage = sorted(
        set(columns)
        & FORBIDDEN_INPUTS
    )

    if leakage:
        raise ValueError(
            f"{variant_name} : "
            f"fuite de données {leakage}"
        )

    print(
        variant_name,
        "->",
        len(columns),
        "entrées",
    )
```

```python
# Cellule 4 — Construction d'un échantillon commun

ALL_ABLATION_INPUTS = sorted(
    {
        column
        for columns in ABLATION_VARIANTS.values()
        for column in columns
    }
)

REQUIRED_COLUMNS = (
    ALL_ABLATION_INPUTS
    + [
        ABLATION_TARGET,
        "alpha_ems_alpha_star_valid",
        TIME_COL,
        "hasPower",
        "SOC_EB",
        "SOC_PB",
    ]
)


def build_common_data(data):
    missing = sorted(
        set(REQUIRED_COLUMNS)
        - set(data.columns)
    )

    if missing:
        raise ValueError(
            f"Colonnes manquantes : {missing}"
        )

    valid = (
        data["alpha_ems_alpha_star_valid"].eq(1)
        & data[ABLATION_TARGET].notna()
        & data[
            ALL_ABLATION_INPUTS
        ].notna().all(axis=1)
    )

    output = (
        data.loc[valid]
        .sort_values(TIME_COL)
        .reset_index(drop=True)
        .copy()
    )

    return output


common_datasets = {
    split_name: build_common_data(data)
    for split_name, data
    in merged_datasets.items()
}

for split_name, data in common_datasets.items():
    print(
        split_name,
        ":",
        len(data),
        "observations communes",
    )
```

```python
# Cellule 5 — Préparation et normalisation par variante


BINARY_COLUMNS = set(
    VARIABLE_SYMBOLIC_COLS
)


def prepare_variant_data(
    variant_name,
    input_columns,
):
    train_data = common_datasets["train"]

    X_train = train_data[
        input_columns
    ].to_numpy(dtype=np.float32)

    y_train = train_data[
        ABLATION_TARGET
    ].to_numpy(dtype=np.float32)

    binary_mask = np.array(
        [
            column in BINARY_COLUMNS
            for column in input_columns
        ],
        dtype=bool,
    )

    raw_mean = X_train.mean(
        axis=0,
        dtype=np.float64,
    ).astype(np.float32)

    raw_std = X_train.std(
        axis=0,
        dtype=np.float64,
    ).astype(np.float32)

    continuous_constant = (
        (~binary_mask)
        & (raw_std < 1e-12)
    )

    if continuous_constant.any():
        bad_columns = np.array(
            input_columns
        )[continuous_constant]

        raise ValueError(
            f"{variant_name} : variables "
            f"continues constantes "
            f"{bad_columns.tolist()}"
        )

    x_mean = np.where(
        binary_mask,
        0.0,
        raw_mean,
    ).astype(np.float32)

    x_std = np.where(
        binary_mask,
        1.0,
        raw_std,
    ).astype(np.float32)

    split_arrays = {}

    for split_name, data in common_datasets.items():
        X = data[
            input_columns
        ].to_numpy(dtype=np.float32)

        y = data[
            ABLATION_TARGET
        ].to_numpy(dtype=np.float32)

        X_normalized = (
            (X - x_mean)
            / x_std
        ).astype(np.float32)

        if not np.isfinite(
            X_normalized
        ).all():
            raise ValueError(
                f"{variant_name}/{split_name} : "
                "valeurs non finies."
            )

        split_arrays[split_name] = {
            "X": X_normalized,
            "y": y,
        }

    scaler_file = (
        MODELS_DIR
        / f"ablation_{variant_name}_scaler.npz"
    )

    np.savez(
        scaler_file,
        x_mean=x_mean,
        x_std=x_std,
        input_cols=np.array(
            input_columns
        ),
        binary_mask=binary_mask,
    )

    return {
        "input_columns": input_columns,
        "x_mean": x_mean,
        "x_std": x_std,
        "splits": split_arrays,
        "scaler_file": scaler_file,
    }


ablation_data = {
    variant_name: prepare_variant_data(
        variant_name,
        input_columns,
    )
    for variant_name, input_columns
    in ABLATION_VARIANTS.items()
}
```

```python
# Cellule 6 — Dataset et architecture commune

class AblationDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X)
        self.y = torch.from_numpy(y)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, index):
        return (
            self.X[index],
            self.y[index],
        )


class DirectAlphaMLP(nn.Module):
    def __init__(
        self,
        n_inputs,
        hidden_1,
        hidden_2,
        dropout,
    ):
        super().__init__()

        self.network = nn.Sequential(
            nn.Linear(
                n_inputs,
                hidden_1,
            ),
            nn.ReLU(),
            nn.Dropout(dropout),

            nn.Linear(
                hidden_1,
                hidden_2,
            ),
            nn.ReLU(),

            nn.Linear(
                hidden_2,
                1,
            ),
        )

    def forward(self, x):
        return torch.sigmoid(
            self.network(x)
        ).squeeze(-1)


def count_parameters(model):
    return sum(
        parameter.numel()
        for parameter in model.parameters()
        if parameter.requires_grad
    )
```

```python
# Cellule 7 — Fonctions de métriques


def regression_mae(
    y_true,
    y_pred,
):
    return float(
        np.mean(
            np.abs(
                y_true - y_pred
            )
        )
    )


def regression_rmse(
    y_true,
    y_pred,
):
    return float(
        np.sqrt(
            np.mean(
                (
                    y_true - y_pred
                ) ** 2
            )
        )
    )


def regression_r2(
    y_true,
    y_pred,
):
    denominator = np.sum(
        (
            y_true
            - np.mean(y_true)
        ) ** 2
    )

    if np.isclose(
        denominator,
        0.0,
    ):
        return np.nan

    return float(
        1.0
        - np.sum(
            (
                y_true - y_pred
            ) ** 2
        )
        / denominator
    )


def predict_array(
    model,
    X,
    batch_size=512,
):
    model.eval()

    predictions = []

    with torch.no_grad():
        for start in range(
            0,
            len(X),
            batch_size,
        ):
            batch = torch.from_numpy(
                X[
                    start:
                    start + batch_size
                ]
            ).to(
                DEVICE,
                non_blocking=True,
            )

            predictions.append(
                model(batch)
                .cpu()
                .numpy()
            )

    return np.concatenate(
        predictions,
        axis=0,
    )
```

```python
# Cellule 8 — Entraînement d’une variante


def train_ablation_model(
    variant_name,
    variant_data,
    seed,
):
    torch.manual_seed(seed)
    np.random.seed(seed)

    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

    train_arrays = variant_data[
        "splits"
    ]["train"]

    val_arrays = variant_data[
        "splits"
    ]["validation"]

    generator = torch.Generator()
    generator.manual_seed(seed)

    train_loader = DataLoader(
        AblationDataset(
            train_arrays["X"],
            train_arrays["y"],
        ),
        batch_size=ABLATION_BATCH_SIZE,
        shuffle=True,
        generator=generator,
        num_workers=ABLATION_NUM_WORKERS,
        pin_memory=ABLATION_PIN_MEMORY,
    )

    val_loader = DataLoader(
        AblationDataset(
            val_arrays["X"],
            val_arrays["y"],
        ),
        batch_size=ABLATION_BATCH_SIZE,
        shuffle=False,
        num_workers=ABLATION_NUM_WORKERS,
        pin_memory=ABLATION_PIN_MEMORY,
    )

    model = DirectAlphaMLP(
        n_inputs=len(
            variant_data[
                "input_columns"
            ]
        ),
        hidden_1=ABLATION_HIDDEN_1,
        hidden_2=ABLATION_HIDDEN_2,
        dropout=ABLATION_DROPOUT,
    ).to(DEVICE)

    criterion = nn.MSELoss()

    optimizer = torch.optim.Adam(
        model.parameters(),
        lr=ABLATION_LEARNING_RATE,
    )

    best_val_loss = float("inf")
    best_epoch = -1
    patience_count = 0
    best_state = None

    history_rows = []

    for epoch in range(
        ABLATION_EPOCHS
    ):
        model.train()

        train_loss_sum = 0.0

        for xb, yb in train_loader:
            xb = xb.to(
                DEVICE,
                non_blocking=True,
            )

            yb = yb.to(
                DEVICE,
                non_blocking=True,
            )

            optimizer.zero_grad(
                set_to_none=True
            )

            prediction = model(xb)

            loss = criterion(
                prediction,
                yb,
            )

            if not torch.isfinite(loss):
                raise ValueError(
                    f"{variant_name} : "
                    "perte non finie."
                )

            loss.backward()

            torch.nn.utils.clip_grad_norm_(
                model.parameters(),
                max_norm=1.0,
            )

            optimizer.step()

            train_loss_sum += (
                loss.item()
                * xb.size(0)
            )

        train_loss = (
            train_loss_sum
            / len(train_loader.dataset)
        )

        model.eval()

        val_loss_sum = 0.0

        with torch.no_grad():
            for xb, yb in val_loader:
                xb = xb.to(
                    DEVICE,
                    non_blocking=True,
                )

                yb = yb.to(
                    DEVICE,
                    non_blocking=True,
                )

                val_loss_sum += (
                    criterion(
                        model(xb),
                        yb,
                    ).item()
                    * xb.size(0)
                )

        val_loss = (
            val_loss_sum
            / len(val_loader.dataset)
        )

        history_rows.append({
            "epoch": epoch + 1,
            "train_loss": train_loss,
            "val_loss": val_loss,
        })

        if (
            val_loss
            < best_val_loss - 1e-6
        ):
            best_val_loss = val_loss
            best_epoch = epoch
            patience_count = 0

            best_state = {
                key: value.detach()
                .cpu()
                .clone()
                for key, value
                in model.state_dict().items()
            }

        else:
            patience_count += 1

        if (
            patience_count
            >= ABLATION_PATIENCE
        ):
            break

    if best_state is None:
        raise RuntimeError(
            f"{variant_name} : "
            "aucun checkpoint valide."
        )

    model.load_state_dict(
        best_state
    )

    checkpoint_file = (
        MODELS_DIR
        / (
            f"ablation_{variant_name}"
            f"_seed_{seed}.pt"
        )
    )

    torch.save(
        best_state,
        checkpoint_file,
    )

    history_df = pd.DataFrame(
        history_rows
    )

    history_df.to_csv(
        TABLES_DIR
        / (
            f"ablation_history_"
            f"{variant_name}_seed_{seed}.csv"
        ),
        index=False,
    )

    return {
        "state_dict": best_state,
        "checkpoint": checkpoint_file,
        "best_epoch": best_epoch + 1,
        "best_val_loss": best_val_loss,
        "n_parameters": count_parameters(
            model
        ),
    }
```

```python
# Cellule 9 — Entraînement des cinq variantes


trained_runs = {}

for variant_name, variant_data in ablation_data.items():
    for seed in ABLATION_SEEDS:
        print(
            "\nEntraînement :",
            variant_name,
            "| seed :",
            seed,
        )

        run = train_ablation_model(
            variant_name,
            variant_data,
            seed,
        )

        trained_runs[
            (variant_name, seed)
        ] = run

        print(
            "Meilleure époque :",
            run["best_epoch"],
            "| val MSE :",
            run["best_val_loss"],
        )
```

```python
# Cellule 10 — Prédictions et métriques de régression

prediction_cache = {}
regression_rows = []

for (
    variant_name,
    seed,
), run in trained_runs.items():

    variant_data = ablation_data[
        variant_name
    ]

    model = DirectAlphaMLP(
        n_inputs=len(
            variant_data[
                "input_columns"
            ]
        ),
        hidden_1=ABLATION_HIDDEN_1,
        hidden_2=ABLATION_HIDDEN_2,
        dropout=ABLATION_DROPOUT,
    ).to(DEVICE)

    model.load_state_dict(
        run["state_dict"]
    )

    for split_name in [
        "train",
        "validation",
        "test",
    ]:
        arrays = variant_data[
            "splits"
        ][split_name]

        start_time = time.perf_counter()

        prediction = predict_array(
            model,
            arrays["X"],
        )

        elapsed = (
            time.perf_counter()
            - start_time
        )

        prediction_cache[
            (
                variant_name,
                seed,
                split_name,
            )
        ] = prediction

        regression_rows.append({
            "strategy": variant_name,
            "seed": seed,
            "split": split_name,
            "mae": regression_mae(
                arrays["y"],
                prediction,
            ),
            "rmse": regression_rmse(
                arrays["y"],
                prediction,
            ),
            "r2": regression_r2(
                arrays["y"],
                prediction,
            ),
            "inference_ms_per_sample": (
                1000.0
                * elapsed
                / len(prediction)
            ),
            "n_parameters": run[
                "n_parameters"
            ],
            "best_epoch": run[
                "best_epoch"
            ],
        })

regression_detail_df = pd.DataFrame(
    regression_rows
)

regression_detail_df.to_csv(
    TABLES_DIR
    / "ablation_regression_detail.csv",
    index=False,
)

display(
    regression_detail_df[
        regression_detail_df[
            "split"
        ].eq("test")
    ]
)
```

```python
# Cellule 11 — Résumé de régression


regression_summary_df = (
    regression_detail_df
    .groupby(
        [
            "strategy",
            "split",
        ],
        as_index=False,
    )
    .agg(
        mae_mean=(
            "mae",
            "mean",
        ),
        mae_std=(
            "mae",
            "std",
        ),
        rmse_mean=(
            "rmse",
            "mean",
        ),
        rmse_std=(
            "rmse",
            "std",
        ),
        r2_mean=(
            "r2",
            "mean",
        ),
        r2_std=(
            "r2",
            "std",
        ),
        inference_ms_mean=(
            "inference_ms_per_sample",
            "mean",
        ),
        parameters=(
            "n_parameters",
            "first",
        ),
    )
)

regression_summary_df.to_csv(
    TABLES_DIR
    / "ablation_regression_summary.csv",
    index=False,
)

display(
    regression_summary_df[
        regression_summary_df[
            "split"
        ].eq("test")
    ].sort_values(
        "mae_mean"
    )
)
```

```python
# Cellule 12 — Évaluation séquentielle du coût physique


evaluate_signature = inspect.signature(
    evaluate_alpha_row
)

EVALUATE_USES_PREVIOUS_ALPHA = (
    "alpha_prev"
    in evaluate_signature.parameters
    or len(
        evaluate_signature.parameters
    ) >= 5
)

print(
    "Continuité utilisée dans "
    "evaluate_alpha_row :",
    EVALUATE_USES_PREVIOUS_ALPHA,
)


def evaluate_alpha_sequence(
    data,
    alpha_values,
):
    n = len(data)

    feasible = np.zeros(
        n,
        dtype=np.int8,
    )

    cost = np.full(
        n,
        np.nan,
    )

    previous_alpha = None

    p_values = data[
        "hasPower"
    ].to_numpy(dtype=float)

    soc_eb_values = data[
        "SOC_EB"
    ].to_numpy(dtype=float)

    soc_pb_values = data[
        "SOC_PB"
    ].to_numpy(dtype=float)

    for index in range(n):
        alpha_value = float(
            alpha_values[index]
        )

        p_dem = float(
            p_values[index]
        )

        soc_eb = float(
            soc_eb_values[index]
        )

        soc_pb = float(
            soc_pb_values[index]
        )

        if EVALUATE_USES_PREVIOUS_ALPHA:
            result = evaluate_alpha_row(
                alpha_value,
                p_dem,
                soc_eb,
                soc_pb,
                previous_alpha,
            )
        else:
            result = evaluate_alpha_row(
                alpha_value,
                p_dem,
                soc_eb,
                soc_pb,
            )

        feasible[index] = int(
            result["feasible"]
        )

        cost[index] = result["cost"]

        if (
            np.isfinite(alpha_value)
            and abs(p_dem)
            > EPS_POWER_W
        ):
            previous_alpha = alpha_value

    return {
        "feasible": feasible,
        "cost": cost,
    }
```

```python
# Cellule 13 — Charger le MLP-NS résiduel actuel


RESIDUAL_PREDICTION_FILE = (
    PREDICTIONS_DIR
    / "predictions_mlp_neurosymbolique.csv"
)

residual_alpha_test = None

if RESIDUAL_PREDICTION_FILE.exists():
    residual_df = pd.read_csv(
        RESIDUAL_PREDICTION_FILE
    )

    if (
        TIME_COL in residual_df.columns
        and "alpha_ems_mlp_neurosymbolic"
        in residual_df.columns
    ):
        if residual_df[
            TIME_COL
        ].duplicated().any():
            raise ValueError(
                "Temps dupliqués dans les "
                "prédictions MLP-NS."
            )

        residual_series = (
            residual_df
            .set_index(TIME_COL)[
                "alpha_ems_mlp_neurosymbolic"
            ]
        )

        residual_alpha_test = (
            common_datasets["test"][
                TIME_COL
            ]
            .map(residual_series)
            .to_numpy(dtype=float)
        )

        if not np.isfinite(
            residual_alpha_test
        ).all():
            print(
                "MLP-NS résiduel non aligné "
                "sur toutes les lignes communes."
            )

            residual_alpha_test = None

if residual_alpha_test is not None:
    print(
        "MLP-NS résiduel chargé :",
        len(residual_alpha_test),
        "prédictions",
    )
else:
    print(
        "MLP-NS résiduel non disponible."
    )
```

```python
# Cellule 14 — Comparaison physique sur validation et test
physical_rows = []

for split_name in ["validation", "test"]:
    data = common_datasets[split_name]

    valid_demand = data["hasPower"].abs().to_numpy() > EPS_POWER_W
    alpha_star_values = data["alpha_ems_alpha_star"].to_numpy(dtype=float)

    for seed in ABLATION_SEEDS:
        strategy_alphas = {"Alpha-star": alpha_star_values}

        for variant_name in ABLATION_VARIANTS:
            strategy_alphas[variant_name] = prediction_cache[(variant_name, seed, split_name)]

        if split_name == "test" and residual_alpha_test is not None:
            strategy_alphas["MLP_neurosymbolique_residuel"] = residual_alpha_test

        evaluation = {
            strategy_name: evaluate_alpha_sequence(data, alpha_values)
            for (strategy_name, alpha_values) in strategy_alphas.items()
        }

        common_feasible = valid_demand.copy()
        for strategy_name in evaluation:
            common_feasible &= (evaluation[strategy_name]["feasible"] == 1)

        if not common_feasible.any():
            raise ValueError(f"Aucun cas commun faisable pour {split_name}, seed={seed}.")

        common_feasible_ratio = common_feasible.sum() / valid_demand.sum()
        if common_feasible_ratio < 0.5:
            print(f"ATTENTION [{split_name}, seed={seed}] : seulement {100*common_feasible_ratio:.1f}% "
                  f"des lignes valides sont communes a toutes les strategies -- comparaison a interpreter avec prudence.")

        alpha_star_cost = evaluation["Alpha-star"]["cost"]

        for (strategy_name, alpha_values) in strategy_alphas.items():
            strategy_eval = evaluation[strategy_name]
            strategy_cost = strategy_eval["cost"]

            physical_rows.append({
                "split": split_name,
                "seed": seed,
                "strategy": strategy_name,
                "valid_demand_samples": int(valid_demand.sum()),
                "common_feasible_samples": int(common_feasible.sum()),
                "common_feasible_percent": float(100.0 * common_feasible.sum() / valid_demand.sum()),
                "feasibility_percent": float(100.0 * strategy_eval["feasible"][valid_demand].mean()),
                "mean_alpha": float(np.mean(alpha_values[valid_demand])),
                "mean_cost_common": float(np.mean(strategy_cost[common_feasible])),
                "mean_regret_common": float(np.mean(strategy_cost[common_feasible] - alpha_star_cost[common_feasible])),
            })

physical_detail_df = pd.DataFrame(physical_rows)
physical_detail_df.to_csv(TABLES_DIR / "ablation_physical_detail.csv", index=False)

display(physical_detail_df[physical_detail_df["split"].eq("test")])
```

```python
# Cellule 15 — Résumé des performances physiques

physical_summary_df = (
    physical_detail_df
    .groupby(
        [
            "strategy",
            "split",
        ],
        as_index=False,
    )
    .agg(
        feasibility_mean=(
            "feasibility_percent",
            "mean",
        ),
        feasibility_std=(
            "feasibility_percent",
            "std",
        ),
        cost_mean=(
            "mean_cost_common",
            "mean",
        ),
        cost_std=(
            "mean_cost_common",
            "std",
        ),
        regret_mean=(
            "mean_regret_common",
            "mean",
        ),
        regret_std=(
            "mean_regret_common",
            "std",
        ),
        common_samples=(
            "common_feasible_samples",
            "min",
        ),
    )
)

physical_summary_df.to_csv(
    TABLES_DIR
    / "ablation_physical_summary.csv",
    index=False,
)

display(
    physical_summary_df[
        physical_summary_df[
            "split"
        ].eq("test")
    ].sort_values(
        "cost_mean"
    )
)
```

```python
# Cellule 16 — Contribution de chaque composante par rapport au MLP simple

test_regression = (
    regression_summary_df[
        regression_summary_df[
            "split"
        ].eq("test")
    ]
    .set_index("strategy")
)

test_physical = (
    physical_summary_df[
        physical_summary_df[
            "split"
        ].eq("test")
        & physical_summary_df[
            "strategy"
        ].isin(
            ABLATION_VARIANTS.keys()
        )
    ]
    .set_index("strategy")
)

baseline_name = (
    "MLP"
)

contribution_rows = []

for strategy_name in (
    ABLATION_VARIANTS
):
    contribution_rows.append({
        "strategy": strategy_name,
        "test_mae": float(
            test_regression.loc[
                strategy_name,
                "mae_mean",
            ]
        ),
        "delta_mae_vs_baseline": float(
            test_regression.loc[
                strategy_name,
                "mae_mean",
            ]
            - test_regression.loc[
                baseline_name,
                "mae_mean",
            ]
        ),
        "test_cost": float(
            test_physical.loc[
                strategy_name,
                "cost_mean",
            ]
        ),
        "delta_cost_vs_baseline": float(
            test_physical.loc[
                strategy_name,
                "cost_mean",
            ]
            - test_physical.loc[
                baseline_name,
                "cost_mean",
            ]
        ),
        "test_feasibility_percent": float(
            test_physical.loc[
                strategy_name,
                "feasibility_mean",
            ]
        ),
    })

contribution_df = pd.DataFrame(
    contribution_rows
).sort_values(
    "test_cost"
)

display(contribution_df)

contribution_df.to_csv(
    TABLES_DIR
    / "ablation_component_contributions.csv",
    index=False,
)
```

```python
# Cellule 17 — Sélection de la meilleure architecture sur validation

validation_ranking = (
    physical_summary_df[
        physical_summary_df[
            "split"
        ].eq("validation")
        & physical_summary_df[
            "strategy"
        ].isin(
            ABLATION_VARIANTS.keys()
        )
    ]
    .sort_values(
        [
            "cost_mean",
            "regret_mean",
        ]
    )
    .reset_index(drop=True)
)

display(validation_ranking)

best_variant = str(
    validation_ranking.loc[
        0,
        "strategy",
    ]
)

best_variant_runs = (
    physical_detail_df[
        physical_detail_df[
            "split"
        ].eq("validation")
        & physical_detail_df[
            "strategy"
        ].eq(best_variant)
    ]
    .sort_values(
        "mean_cost_common"
    )
)

best_seed = int(
    best_variant_runs.iloc[0][
        "seed"
    ]
)

selection = {
    "selected_on": "validation",
    "criterion": (
        "minimum mean physical cost "
        "on common feasible samples"
    ),
    "best_variant": best_variant,
    "best_seed": best_seed,
    "input_columns": (
        ABLATION_VARIANTS[
            best_variant
        ]
    ),
    "checkpoint": str(
        trained_runs[
            (
                best_variant,
                best_seed,
            )
        ]["checkpoint"]
    ),
    "scaler": str(
        ablation_data[
            best_variant
        ]["scaler_file"]
    ),
}

selection_file = (
    MODELS_DIR
    / "ablation_selected_model.json"
)

with open(
    selection_file,
    "w",
    encoding="utf-8",
) as file:
    json.dump(
        selection,
        file,
        indent=2,
        ensure_ascii=False,
    )

print(
    "Architecture sélectionnée :",
    best_variant,
)

print("Seed sélectionné :", best_seed)
print("Configuration :", selection_file)
```
