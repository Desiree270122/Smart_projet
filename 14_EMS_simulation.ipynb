```python
%run "./13_EMS_safety_filter.ipynb"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- fonctions d'epaulement propres (corrige le bug trapmf aux bornes exactes) ---
def shoulder_low(x, b, c):
    x = np.asarray(x, dtype=float)
    y = np.where(x <= b, 1.0, 0.0)
    mid = (x > b) & (x < c)
    y = np.where(mid, (c - x) / (c - b), y)
    return y

def shoulder_high(x, b, c):
    x = np.asarray(x, dtype=float)
    y = np.where(x >= c, 1.0, 0.0)
    mid = (x > b) & (x < c)
    y = np.where(mid, (x - b) / (c - b), y)
    return y

def trapmf(x, a, b, c, d):
    x = np.asarray(x, dtype=float)
    b = b if b > a else a + 1e-9
    c = c if c < d else d - 1e-9
    y = np.where((x >= b) & (x <= c), 1.0, 0.0)
    left = (x > a) & (x < b); y = np.where(left, (x - a) / (b - a), y)
    right = (x > c) & (x < d); y = np.where(right, (d - x) / (d - c), y)
    return y

def soc_faible(s): return shoulder_low(s, 0.25, SOC_LOW_THRESHOLD)
def soc_moyen(s):  return trapmf(s, 0.25, 0.35, 0.65, 0.75)
def soc_eleve(s):  return shoulder_high(s, 0.70, 0.80)

def p_forte_recharge(p):  return shoulder_low(p, P_EB_MIN_W - 2000, P_EB_MIN_W)
def p_faible_recharge(p): return trapmf(p, P_EB_MIN_W, P_EB_MIN_W * 0.55, -2 * EPS_POWER_W, -EPS_POWER_W)
def p_nulle(p):            return trapmf(p, -2 * EPS_POWER_W, -EPS_POWER_W, EPS_POWER_W, 2 * EPS_POWER_W)
def p_faible_traction(p): return trapmf(p, EPS_POWER_W, 2 * EPS_POWER_W, P_EB_MAX_W * 0.6, P_EB_MAX_W)
def p_forte_traction(p):  return shoulder_high(p, P_EB_MAX_W * 0.6, P_EB_MAX_W)

def a_freinage(a): return shoulder_low(a, -1.2, -0.2)
def a_stable(a):   return trapmf(a, -0.4, -0.1, 0.1, 0.4)

def alpha_fuzzy_calc(soc_eb, soc_pb, pdem, accel):
    eb_faible = soc_faible(soc_eb); pb_faible = soc_faible(soc_pb)
    pb_moyen = soc_moyen(soc_pb); pb_eleve = soc_eleve(soc_pb)
    p_fr, p_fbr = p_forte_recharge(pdem), p_faible_recharge(pdem)
    p_n, p_fbt, p_ft = p_nulle(pdem), p_faible_traction(pdem), p_forte_traction(pdem)
    a_frein, a_stab = a_freinage(accel), a_stable(accel)
    traction = np.maximum(p_fbt, p_ft); recharge = np.maximum(np.maximum(p_fr, p_fbr), a_frein)
    pb_dispo = np.maximum(pb_moyen, pb_eleve); pb_rechargeable = np.maximum(pb_faible, pb_moyen)
    w1, c1 = np.minimum(pb_faible, traction), 0.20
    w2, c2 = np.minimum(np.minimum(eb_faible, pb_dispo), traction), 0.75
    w3, c3 = np.minimum(p_ft, pb_dispo), 0.75
    w4, c4 = np.minimum(p_n, a_stab), 0.20
    w5, c5 = np.minimum(recharge, pb_rechargeable), 0.75
    w7, c7 = np.minimum(eb_faible, pb_faible), 0.50
    ws = np.stack([w1, w2, w3, w4, w5, w7]); cs = np.array([c1, c2, c3, c4, c5, c7])
    wsum = ws.sum(axis=0)
    return np.where(wsum > 1e-6, (ws * cs[:, None]).sum(axis=0) / np.maximum(wsum, 1e-9), 0.30)
```

```python
# Cellule 3 : safety_filter etendu (P_EB_final, P_PB_final, P_regen_curtailed, P_unserved)
def diagnose_violation(alpha_req, p_dem, soc_eb, soc_pb):
    reasons = []
    if alpha_req < 0 or alpha_req > 1: reasons.append('alpha_hors_0_1')
    p_pb = alpha_req * p_dem; p_eb = p_dem - p_pb
    p_conv = estimate_p_conv(np.array([p_eb]))[0]
    if p_eb < P_EB_MIN_W or p_eb > P_EB_MAX_W: reasons.append('P_EB_hors_limites')
    if p_pb < P_PB_MIN_W or p_pb > P_PB_MAX_W: reasons.append('P_PB_hors_limites')
    if p_conv < P_CONV_MIN_W or p_conv > P_CONV_MAX_W: reasons.append('P_conv_hors_limites')
    if soc_eb <= SOC_EB_MIN + SOC_TOL and p_eb > EPS_POWER_W: reasons.append('EB_protection_SOC_bas')
    if soc_pb <= SOC_PB_MIN + SOC_TOL and p_pb > EPS_POWER_W: reasons.append('PB_protection_SOC_bas')
    i_eb, i_pb = p_eb / V_EB_PACK_NOM, p_pb / V_PB_PACK_NOM
    soc_eb_next = soc_eb - i_eb * DT_OPT_SECONDS / (3600.0 * CAPACITY_EB_AH)
    soc_pb_next = soc_pb - i_pb * DT_OPT_SECONDS / (3600.0 * CAPACITY_PB_AH)
    if soc_eb_next < SOC_EB_MIN - SOC_TOL or soc_eb_next > SOC_EB_MAX + SOC_TOL: reasons.append('SOC_EB_projete_hors_bornes')
    if soc_pb_next < SOC_PB_MIN - SOC_TOL or soc_pb_next > SOC_PB_MAX + SOC_TOL: reasons.append('SOC_PB_projete_hors_bornes')
    return reasons


def eb_priority_alpha_single(p_dem, soc_eb):
    if abs(p_dem) <= EPS_POWER_W: return 0.5
    if p_dem < P_EB_MIN_W: p_eb = P_EB_MIN_W
    elif p_dem < 0: p_eb = p_dem
    elif soc_eb <= SOC_EB_MIN: p_eb = 0.0
    elif p_dem <= P_EB_MAX_W: p_eb = p_dem
    else: p_eb = P_EB_MAX_W
    p_pb = p_dem - p_eb
    return float(np.clip(p_pb / p_dem, 0.0, 1.0))


def safety_filter(alpha_req, p_dem, soc_eb, soc_pb, alpha_prev=None):
    if abs(p_dem) <= EPS_POWER_W:
        return {'alpha_requested': alpha_req, 'alpha_final': alpha_req,
                'P_EB_final': 0.0, 'P_PB_final': 0.0, 'P_regen_curtailed': 0.0, 'P_unserved': 0.0,
                'correction_applied': False, 'violated_constraint': 'aucune', 'explanation': 'demande nulle'}

    metrics = candidate_metrics(ALPHA_GRID, p_dem, soc_eb, soc_pb, alpha_prev)
    feasible_idx = np.flatnonzero(metrics['feasible'])

    if len(feasible_idx) == 0:
        alpha_fallback = eb_priority_alpha_single(p_dem, soc_eb)
        p_pb_raw = alpha_fallback * p_dem
        p_eb_raw = p_dem - p_pb_raw
        p_eb_final = float(np.clip(p_eb_raw, P_EB_MIN_W, P_EB_MAX_W))
        p_pb_final = float(np.clip(p_pb_raw, P_PB_MIN_W, P_PB_MAX_W))
        residual = p_dem - (p_eb_final + p_pb_final)
        p_regen_curtailed = float(residual) if p_dem < 0 else 0.0
        p_unserved = float(residual) if p_dem >= 0 else 0.0
        return {'alpha_requested': alpha_req, 'alpha_final': alpha_fallback,
                'P_EB_final': p_eb_final, 'P_PB_final': p_pb_final,
                'P_regen_curtailed': p_regen_curtailed, 'P_unserved': p_unserved,
                'correction_applied': True, 'violated_constraint': 'aucune_solution_faisable',
                'explanation': 'Aucune repartition faisable ; repli EB-priority avec limites absolues, residu trace explicitement.'}

    feasible_alphas = ALPHA_GRID[feasible_idx]
    nearest = feasible_idx[np.argmin(np.abs(feasible_alphas - alpha_req))]
    alpha_final = float(ALPHA_GRID[nearest])
    p_pb_final = alpha_final * p_dem
    p_eb_final = p_dem - p_pb_final
    corrected = not np.isclose(alpha_final, alpha_req, atol=ALPHA_GRID_STEP * 1.1)
    reasons = diagnose_violation(alpha_req, p_dem, soc_eb, soc_pb) if corrected else []

    return {'alpha_requested': alpha_req, 'alpha_final': alpha_final,
            'P_EB_final': float(p_eb_final), 'P_PB_final': float(p_pb_final),
            'P_regen_curtailed': 0.0, 'P_unserved': 0.0,
            'correction_applied': corrected, 'violated_constraint': ','.join(reasons) if reasons else 'aucune',
            'explanation': 'aucune correction necessaire' if not corrected else f"Projection (causes : {', '.join(reasons)})"}
```

```python
# Cellule 4 : mise a jour SOC avec verification avant clip
def update_soc_checked(soc_eb, soc_pb, p_eb, p_pb):
    i_eb, i_pb = p_eb / V_EB_PACK_NOM, p_pb / V_PB_PACK_NOM
    soc_eb_raw = soc_eb - i_eb * DT_OPT_SECONDS / (3600.0 * CAPACITY_EB_AH)
    soc_pb_raw = soc_pb - i_pb * DT_OPT_SECONDS / (3600.0 * CAPACITY_PB_AH)

    large_tol = 0.02  # au-dela, ce n'est plus une simple saturation numerique
    if soc_eb_raw < SOC_EB_MIN - large_tol or soc_eb_raw > SOC_EB_MAX + large_tol:
        raise ValueError(f"SOC_EB largement hors limites : {soc_eb_raw}")
    if soc_pb_raw < SOC_PB_MIN - large_tol or soc_pb_raw > SOC_PB_MAX + large_tol:
        raise ValueError(f"SOC_PB largement hors limites : {soc_pb_raw}")

    return float(np.clip(soc_eb_raw, SOC_EB_MIN, SOC_EB_MAX)), float(np.clip(soc_pb_raw, SOC_PB_MIN, SOC_PB_MAX))
```

```python
# Cellule 5 : moteur de simulation
def simulate(p_dem_seq, alpha_func, soc_eb0, soc_pb0):
    n = len(p_dem_seq)
    soc_eb, soc_pb = soc_eb0, soc_pb0
    alpha_prev = None

    traj = {k: np.zeros(n) for k in
            ['SOC_EB', 'SOC_PB', 'P_EB', 'P_PB', 'alpha_final', 'P_regen_curtailed', 'P_unserved', 'balance_error_W']}
    traj['corrected'] = np.zeros(n, dtype=bool)

    for t in range(n):
        p_dem = p_dem_seq[t]
        traj['SOC_EB'][t] = soc_eb
        traj['SOC_PB'][t] = soc_pb

        alpha_req = alpha_func(p_dem, soc_eb, soc_pb, t)
        filtered = safety_filter(alpha_req, p_dem, soc_eb, soc_pb, alpha_prev)

        p_eb, p_pb = filtered['P_EB_final'], filtered['P_PB_final']
        curtailed, unserved = filtered['P_regen_curtailed'], filtered['P_unserved']

        traj['P_EB'][t], traj['P_PB'][t] = p_eb, p_pb
        traj['alpha_final'][t] = filtered['alpha_final']
        traj['P_regen_curtailed'][t], traj['P_unserved'][t] = curtailed, unserved
        traj['balance_error_W'][t] = p_dem - p_eb - p_pb - curtailed - unserved
        traj['corrected'][t] = filtered['correction_applied']

        soc_eb, soc_pb = update_soc_checked(soc_eb, soc_pb, p_eb, p_pb)
        if abs(p_dem) > EPS_POWER_W:
            alpha_prev = filtered['alpha_final']

    traj['SOC_EB_final_after'] = soc_eb
    traj['SOC_PB_final_after'] = soc_pb
    traj['n_corrections'] = int(traj['corrected'].sum())
    traj['curtailed_energy_Wh'] = float(np.sum(np.abs(traj['P_regen_curtailed'])) * DT_OPT_SECONDS / 3600.0)
    traj['unserved_energy_Wh'] = float(np.sum(traj['P_unserved']) * DT_OPT_SECONDS / 3600.0)
    return traj
```

```python
# Cellule 6 : strategies (I_EB causal pour l'approche complete)
CONVERTER_RISK_THRESHOLD = globals().get('CONVERTER_RISK_THRESHOLD', 0.85)

def historique_func(p_dem, soc_eb, soc_pb, t):
    return eb_priority_alpha_single(p_dem, soc_eb)

def floue_func(p_dem, soc_eb, soc_pb, t):
    return float(alpha_fuzzy_calc(np.array([soc_eb]), np.array([soc_pb]), np.array([p_dem]), np.array([accel_seq[t]]))[0])

def mlp_simple_func(p_dem, soc_eb, soc_pb, t):
    feats = {'SOC_EB': soc_eb, 'SOC_PB': soc_pb, 'hasPower': p_dem, 'speed': speed_seq[t], 'hasAcceleration': accel_seq[t]}
    x = np.array([feats[c] for c in mlp_simple_cols], dtype=np.float32)
    xn = (x - mlp_x_mean) / mlp_x_std
    with torch.no_grad():
        a = mlp_simple_model(torch.from_numpy(xn).unsqueeze(0).to(DEVICE)).item()
    return a


class RollingBuffer:
    def __init__(self, window, feature_names, init_rows_df):
        self.window = window
        self.feature_names = feature_names
        self.buffer = init_rows_df[feature_names].to_numpy(dtype=np.float32)[-window:].copy()
    def push(self, feat_dict):
        new_row = np.array([feat_dict[f] for f in self.feature_names], dtype=np.float32)
        self.buffer = np.vstack([self.buffer[1:], new_row])
    def get(self): return self.buffer.copy()


def make_complete_func():
    buffer = RollingBuffer(WINDOW, lstm_input_cols_actual, val_tail)
    state = {'prev_p_eb': float(val_tail['P_EB'].iloc[-1]) if 'P_EB' in val_tail.columns else 0.0}

    def complete_func(p_dem, soc_eb, soc_pb, t):
        eb_avail = float(soc_eb > SOC_EB_MIN); pb_avail = float(soc_pb > SOC_PB_MIN)
        eb_low = float(soc_eb <= SOC_LOW_THRESHOLD); pb_low = float(soc_pb <= SOC_LOW_THRESHOLD)
        high_pow = float(abs(p_dem) >= HIGH_POWER_THRESHOLD_W)
        regen = float(p_dem < -EPS_POWER_W); zero_pow = float(abs(p_dem) <= EPS_POWER_W)

        alpha_fz = float(alpha_fuzzy_calc(np.array([soc_eb]), np.array([soc_pb]), np.array([p_dem]), np.array([accel_seq[t]]))[0])
        p_eb_provisional = (1.0 - alpha_fz) * p_dem
        conv_est = estimate_p_conv(np.array([p_eb_provisional]))[0]
        conv_risk = float((conv_est > CONVERTER_RISK_THRESHOLD * P_CONV_MAX_W) or (conv_est < CONVERTER_RISK_THRESHOLD * P_CONV_MIN_W))

        i_eb_prev = state['prev_p_eb'] / V_EB_PACK_NOM

        symbolic_row = {
            'speed': speed_seq[t], 'hasPower': p_dem, 'hasAcceleration': accel_seq[t],
            'hasTotalForce': test_full['hasTotalForce'].iloc[t], 'SOC_EB': soc_eb, 'SOC_PB': soc_pb,
            'I_EB': i_eb_prev,
            'EB_available': eb_avail, 'PB_available': pb_avail, 'EB_low_SOC': eb_low, 'PB_low_SOC': pb_low,
            'high_power_demand': high_pow, 'regenerative_braking': regen, 'zero_power_demand': zero_pow,
            'converter_risk': conv_risk,
        }
        buffer.push({k: symbolic_row.get(k, 0.0) for k in lstm_input_cols_actual})

        window_x = ((buffer.get() - lstm_x_mean) / lstm_x_std).astype(np.float32)
        with torch.no_grad():
            lstm_pred = lstm_ns_model(torch.from_numpy(window_x).unsqueeze(0).to(DEVICE)).cpu().numpy()[0]
        lstm_pred = lstm_pred * lstm_y_std + lstm_y_mean

        mlp_ns_feats = {
            'SOC_EB': soc_eb, 'SOC_PB': soc_pb, 'hasPower': p_dem, 'speed': speed_seq[t], 'hasAcceleration': accel_seq[t],
            'Pdem_pred_ns': lstm_pred[0], 'delta_soc_eb_pred_ns': lstm_pred[1], 'delta_soc_pb_pred_ns': lstm_pred[2],
            'alpha_ems_fuzzy_logic': alpha_fz,
            'EB_available': eb_avail, 'PB_available': pb_avail, 'EB_low_SOC': eb_low, 'PB_low_SOC': pb_low,
            'high_power_demand': high_pow, 'regenerative_braking': regen, 'zero_power_demand': zero_pow,
            'converter_risk': conv_risk,
        }
        x_ns = np.array([mlp_ns_feats[c] for c in MLP_NS_INPUT_COLS], dtype=np.float32)
        x_ns_n = norm_x(x_ns.reshape(1, -1))[0]

        with torch.no_grad():
            alpha_mlp_ns, _ = model(torch.from_numpy(x_ns_n).unsqueeze(0).to(DEVICE),
                                     torch.tensor([alpha_fz], dtype=torch.float32).to(DEVICE))
        alpha_mlp_ns = alpha_mlp_ns.item()
        alpha_out = (1 - GAMMA_FUSION) * alpha_fz + GAMMA_FUSION * alpha_mlp_ns

        p_pb_provisional = alpha_out * p_dem
        state['prev_p_eb'] = p_dem - p_pb_provisional
        return alpha_out

    return complete_func, lambda p_eb_final: state.update({'prev_p_eb': p_eb_final})
```

```python
# Cellule 7 : execution
strategies = {'historique': historique_func, 'logique_floue': floue_func, 'mlp_simple': mlp_simple_func}
complete_func, update_prev_eb = make_complete_func()
strategies['neurosymbolique_complet'] = complete_func

results = {}
for name, func in strategies.items():
    print('Simulation :', name)
    results[name] = simulate(p_seq, func, soc_eb0, soc_pb0)
    print('  SOC_EB (meme instant que la derniere ligne reelle) :', results[name]['SOC_EB'][-1],
          '| corrections :', results[name]['n_corrections'],
          '| energie regen perdue (Wh):', round(results[name]['curtailed_energy_Wh'], 2))
```
