```python
# Cellule 3 : A -- bruit des capteurs (perturbation des entrees, evaluation ponctuelle)
NOISE_LEVELS_SENSOR = {
    'faible': {'soc_sigma': 0.001, 'power_pct': 0.01},
    'moyen':  {'soc_sigma': 0.005, 'power_pct': 0.03},
    'fort':   {'soc_sigma': 0.010, 'power_pct': 0.05},
}

def add_sensor_noise(data, soc_sigma, power_pct, rng):
    d = data.copy()
    if 'SOC_EB' in d.columns:
        d['SOC_EB'] = np.clip(d['SOC_EB'] + rng.normal(0, soc_sigma, len(d)), 0.0, 1.0)
    if 'SOC_PB' in d.columns:
        d['SOC_PB'] = np.clip(d['SOC_PB'] + rng.normal(0, soc_sigma, len(d)), 0.0, 1.0)
    if 'hasPower' in d.columns:
        sigma_p = np.maximum(power_pct * np.abs(d['hasPower'].to_numpy()), 10.0)
        d['hasPower'] = d['hasPower'] + rng.normal(0, 1, len(d)) * sigma_p
    return d

def predict_variant(variant_name, data_noisy, seed):
    variant_data = ablation_data[variant_name]
    cols = variant_data['input_columns']
    x_mean, x_std = variant_data['x_mean'], variant_data['x_std']
    X = data_noisy[cols].to_numpy(dtype=np.float32)
    Xn = ((X - x_mean) / x_std).astype(np.float32)

    model = DirectAlphaMLP(len(cols), ABLATION_HIDDEN_1, ABLATION_HIDDEN_2, ABLATION_DROPOUT).to(DEVICE)
    model.load_state_dict(trained_runs[(variant_name, seed)]['state_dict'])
    return predict_array(model, Xn)

test_data = common_datasets['test']
alpha_star_test = test_data['alpha_ems_alpha_star'].to_numpy(dtype=float)
valid_demand_test = test_data['hasPower'].abs().to_numpy() > EPS_POWER_W

sensor_noise_rows = []
for level_name, params in NOISE_LEVELS_SENSOR.items():
    for variant_name in ABLATION_VARIANTS:
        for rep in range(ROBUSTNESS_REPEATS):
            seed = ABLATION_SEEDS[rep % len(ABLATION_SEEDS)]
            noisy_data = add_sensor_noise(test_data, params['soc_sigma'], params['power_pct'], RNG)
            alpha_pred = predict_variant(variant_name, noisy_data, seed)

            ev = evaluate_alpha_sequence(noisy_data, alpha_pred)
            common_ok = valid_demand_test & (ev['feasible'] == 1)

            sensor_noise_rows.append({
                'level': level_name, 'variant': variant_name, 'rep': rep, 'seed': seed,
                'alpha_std_shift': float(np.std(alpha_pred[valid_demand_test] - predict_variant(variant_name, test_data, seed)[valid_demand_test])),
                'mae_vs_alpha_star': float(np.mean(np.abs(alpha_pred[valid_demand_test] - alpha_star_test[valid_demand_test]))),
                'mean_cost': float(np.mean(ev['cost'][common_ok])) if common_ok.any() else np.nan,
                'feasibility_avant_filtre_pct': float(100 * ev['feasible'][valid_demand_test].mean()),
            })

sensor_noise_df = pd.DataFrame(sensor_noise_rows)
sensor_noise_summary = sensor_noise_df.groupby(['level', 'variant'], as_index=False).agg(
    mae_mean=('mae_vs_alpha_star', 'mean'), mae_std=('mae_vs_alpha_star', 'std'),
    cost_mean=('mean_cost', 'mean'), cost_std=('mean_cost', 'std'),
    feasibility_mean=('feasibility_avant_filtre_pct', 'mean'),
)
sensor_noise_summary.to_csv(TABLES_DIR / 'robustesse_bruit_capteurs.csv', index=False)
display(sensor_noise_summary.sort_values(['level', 'cost_mean']))
```

```python
# Cellule 5 : C -- robustesse des etats symboliques (inversion aleatoire, variantes B et E)
SYMBOLIC_FLIP_RATES = [0.01, 0.05, 0.10]

def flip_symbolic(data, flip_rate, rng):
    d = data.copy()
    for col in VARIABLE_SYMBOLIC_COLS:
        flip_mask = rng.random(len(d)) < flip_rate
        d.loc[flip_mask, col] = 1.0 - d.loc[flip_mask, col]
    return d

symbolic_noise_rows = []
for rate in SYMBOLIC_FLIP_RATES:
    for variant_name in ['MLP_symbolique', 'MLP_neurosymbolique']:
        for rep in range(ROBUSTNESS_REPEATS):
            seed = ABLATION_SEEDS[rep % len(ABLATION_SEEDS)]
            flipped_data = flip_symbolic(test_data, rate, RNG)
            alpha_pred = predict_variant(variant_name, flipped_data, seed)
            ev = evaluate_alpha_sequence(flipped_data, alpha_pred)
            common_ok = valid_demand_test & (ev['feasible'] == 1)
            symbolic_noise_rows.append({
                'flip_rate': rate, 'variant': variant_name, 'rep': rep, 'seed': seed,
                'mae_vs_alpha_star': float(np.mean(np.abs(alpha_pred[valid_demand_test] - alpha_star_test[valid_demand_test]))),
                'mean_cost': float(np.mean(ev['cost'][common_ok])) if common_ok.any() else np.nan,
                'feasibility_avant_filtre_pct': float(100 * ev['feasible'][valid_demand_test].mean()),
            })

symbolic_noise_df = pd.DataFrame(symbolic_noise_rows)
symbolic_noise_summary = symbolic_noise_df.groupby(['flip_rate', 'variant'], as_index=False).agg(
    mae_mean=('mae_vs_alpha_star', 'mean'), mae_std=('mae_vs_alpha_star', 'std'),
    cost_mean=('mean_cost', 'mean'), cost_std=('mean_cost', 'std'),
    feasibility_mean=('feasibility_avant_filtre_pct', 'mean'),
)
symbolic_noise_summary.to_csv(TABLES_DIR / 'robustesse_bruit_symbolique.csv', index=False)
display(symbolic_noise_summary.sort_values(['flip_rate', 'cost_mean']))
```

```python
# Cellule 6 : metriques dynamiques sous bruit -- energie rejetee/non satisfaite, ecart de trajectoire SOC
# (boucle fermee : le controleur voit des mesures bruitees, la physique reste exacte)
def eb_priority_alpha_single(p_dem, soc_eb):
    if abs(p_dem) <= EPS_POWER_W: return 0.5
    if p_dem < P_EB_MIN_W: p_eb = P_EB_MIN_W
    elif p_dem < 0: p_eb = p_dem
    elif soc_eb <= SOC_EB_MIN: p_eb = 0.0
    elif p_dem <= P_EB_MAX_W: p_eb = p_dem
    else: p_eb = P_EB_MAX_W
    p_pb = p_dem - p_eb
    return float(np.clip(p_pb / p_dem, 0.0, 1.0))

def safety_filter_dyn(alpha_req, p_dem, soc_eb, soc_pb, alpha_prev=None):
    if abs(p_dem) <= EPS_POWER_W:
        return {'alpha_final': alpha_req, 'P_EB_final': 0.0, 'P_PB_final': 0.0, 'P_regen_curtailed': 0.0, 'P_unserved': 0.0}
    metrics = candidate_metrics(ALPHA_GRID, p_dem, soc_eb, soc_pb, alpha_prev)
    idx = np.flatnonzero(metrics['feasible'])
    if len(idx) == 0:
        a_fb = eb_priority_alpha_single(p_dem, soc_eb)
        p_pb_raw, p_eb_raw = a_fb * p_dem, p_dem - a_fb * p_dem
        p_eb_f = float(np.clip(p_eb_raw, P_EB_MIN_W, P_EB_MAX_W))
        p_pb_f = float(np.clip(p_pb_raw, P_PB_MIN_W, P_PB_MAX_W))
        residual = p_dem - (p_eb_f + p_pb_f)
        return {'alpha_final': a_fb, 'P_EB_final': p_eb_f, 'P_PB_final': p_pb_f,
                'P_regen_curtailed': float(residual) if p_dem < 0 else 0.0,
                'P_unserved': float(residual) if p_dem >= 0 else 0.0}
    feas_a = ALPHA_GRID[idx]
    nearest = idx[np.argmin(np.abs(feas_a - alpha_req))]
    a_f = float(ALPHA_GRID[nearest])
    return {'alpha_final': a_f, 'P_EB_final': float(p_dem - a_f * p_dem), 'P_PB_final': float(a_f * p_dem),
            'P_regen_curtailed': 0.0, 'P_unserved': 0.0}

def update_soc_dyn(soc_eb, soc_pb, p_eb, p_pb):
    i_eb, i_pb = p_eb / V_EB_PACK_NOM, p_pb / V_PB_PACK_NOM
    new_eb = soc_eb - i_eb * DT_OPT_SECONDS / (3600.0 * CAPACITY_EB_AH)
    new_pb = soc_pb - i_pb * DT_OPT_SECONDS / (3600.0 * CAPACITY_PB_AH)
    return float(np.clip(new_eb, 0.0, 1.0)), float(np.clip(new_pb, 0.0, 1.0))

def simulate_noisy(p_dem_seq, soc_eb0, soc_pb0, alpha_func, soc_sigma, power_pct, rng):
    n = len(p_dem_seq)
    soc_eb, soc_pb = soc_eb0, soc_pb0
    alpha_prev = None
    soc_eb_traj = np.zeros(n)
    curtailed_total, unserved_total = 0.0, 0.0

    for t in range(n):
        p_dem = p_dem_seq[t]
        soc_eb_traj[t] = soc_eb

        soc_eb_meas = float(np.clip(soc_eb + rng.normal(0, soc_sigma), 0, 1))
        soc_pb_meas = float(np.clip(soc_pb + rng.normal(0, soc_sigma), 0, 1))
        p_dem_meas = p_dem + rng.normal(0, max(power_pct * abs(p_dem), 10.0))

        alpha_req = alpha_func(p_dem_meas, soc_eb_meas, soc_pb_meas)
        result = safety_filter_dyn(alpha_req, p_dem, soc_eb, soc_pb, alpha_prev)
        curtailed_total += abs(result['P_regen_curtailed']) * DT_OPT_SECONDS / 3600.0
        unserved_total += result['P_unserved'] * DT_OPT_SECONDS / 3600.0

        soc_eb, soc_pb = update_soc_dyn(soc_eb, soc_pb, result['P_EB_final'], result['P_PB_final'])
        if abs(p_dem) > EPS_POWER_W:
            alpha_prev = result['alpha_final']

    return soc_eb_traj, soc_eb, curtailed_total, unserved_total

test_full = merged_datasets['test'].reset_index(drop=True)
p_seq_full = test_full['hasPower'].to_numpy()
soc_eb0_full, soc_pb0_full = float(test_full['SOC_EB'].iloc[0]), float(test_full['SOC_PB'].iloc[0])
clean_traj, _, _, _ = simulate_noisy(p_seq_full, soc_eb0_full, soc_pb0_full,
                                       lambda p, se, sp: eb_priority_alpha_single(p, se), 0.0, 0.0, RNG)

DYNAMIC_REPEATS = 10  # reduit par rapport aux 25 du test ponctuel, pour le temps de calcul
dynamic_rows = []
for level_name, params in NOISE_LEVELS_SENSOR.items():
    for rep in range(DYNAMIC_REPEATS):
        traj, soc_final, curtailed, unserved = simulate_noisy(
            p_seq_full, soc_eb0_full, soc_pb0_full,
            lambda p, se, sp: eb_priority_alpha_single(p, se),
            params['soc_sigma'], params['power_pct'], RNG)
        dynamic_rows.append({
            'level': level_name, 'rep': rep,
            'soc_eb_final': soc_final,
            'ecart_trajectoire_soc_mae': float(np.mean(np.abs(traj - clean_traj))),
            'curtailed_energy_Wh': curtailed, 'unserved_energy_Wh': unserved,
        })

dynamic_df = pd.DataFrame(dynamic_rows)
dynamic_summary = dynamic_df.groupby('level', as_index=False).mean(numeric_only=True)
dynamic_summary.to_csv(TABLES_DIR / 'robustesse_dynamique_historique.csv', index=False)
display(dynamic_summary)
```
