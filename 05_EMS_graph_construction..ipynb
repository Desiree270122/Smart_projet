```python

%run "./03_EMS_baselines.ipynb"

%run "./02_EMS_preprocessing.ipynb"
import numpy as np
import pandas as pd
import torch

from torch_geometric.data import Data

train_file = PROCESSED_DIR / "train_with_baselines.csv"
val_file = PROCESSED_DIR / "validation_with_baselines.csv"
test_file = PROCESSED_DIR / "test_with_baselines.csv"

for path in [train_file, val_file, test_file]:
    if not path.exists():
        raise FileNotFoundError(f"Fichier absent : {path}")

train_df = pd.read_csv(train_file)
val_df = pd.read_csv(val_file)
test_df = pd.read_csv(test_file)

datasets = {
    "train": train_df,
    "validation": val_df,
    "test": test_df,
}

for name, data in datasets.items():
    print(name, data.shape)
```

Vérification des colonnes

```python
required_cols = [
    TIME_COL,
    "SOC_EB",
    "SOC_PB",
    "hasPower",
    "hasAcceleration",
    "alpha_ems_eb_priority",
    "alpha_ems_eb_priority_valid",
]

for name, data in datasets.items():
    missing = sorted(
        set(required_cols) - set(data.columns)
    )

    if missing:
        raise ValueError(
            f"{name} : colonnes absentes {missing}"
        )

    if data[required_cols].isna().all(axis=0).any():
        raise ValueError(
            f"{name} contient une colonne entièrement vide."
        )

print("Colonnes validées.")
```

Création des noeuds et arrets

```python
NODE_INDEX = {
    name: index
    for index, name in enumerate(GNN_NODE_NAMES)
}

UNDIRECTED_EDGES = [
    ("energy_battery", "converter"),
    ("power_battery", "converter"),
    ("converter", "motor"),
    ("energy_battery", "motor"),
    ("motor", "vehicle"),
]

edge_list = []

for source, target in UNDIRECTED_EDGES:
    source_index = NODE_INDEX[source]
    target_index = NODE_INDEX[target]

    edge_list.append(
        (source_index, target_index)
    )

    edge_list.append(
        (target_index, source_index)
    )

edge_index = torch.tensor(
    edge_list,
    dtype=torch.long,
).t().contiguous()

print("Nombre de nœuds :", len(GNN_NODE_NAMES))
print("Nombre d'arêtes :", edge_index.shape[1])
print(edge_index)
```

Parametres des noeuds

```python
STATIC_FEATURES = {
    "energy_battery": {
        "current_discharge_max_a": I_EB_MAX_A,
        "capacity_ah": CAPACITY_EB_AH,
        "power_discharge_max_w": P_EB_MAX_W,
        "power_charge_max_abs_w": abs(P_EB_MIN_W),
    },
    "power_battery": {
        "current_discharge_max_a": I_PB_MAX_A,
        "capacity_ah": CAPACITY_PB_AH,
        "power_discharge_max_w": P_PB_MAX_W,
        "power_charge_max_abs_w": abs(P_PB_MIN_W),
    },
    "converter": {
        "current_discharge_max_a": (
            P_CONV_MAX_W / V_HESS_BUS_NOM
        ),
        "capacity_ah": 0.0,
        "power_discharge_max_w": P_CONV_MAX_W,
        "power_charge_max_abs_w": abs(P_CONV_MIN_W),
    },
    "motor": {
        "current_discharge_max_a": 0.0,
        "capacity_ah": 0.0,
        "power_discharge_max_w": 0.0,
        "power_charge_max_abs_w": 0.0,
    },
    "vehicle": {
        "current_discharge_max_a": 0.0,
        "capacity_ah": 0.0,
        "power_discharge_max_w": 0.0,
        "power_charge_max_abs_w": 0.0,
    },
}
```

Construction des caractéristiques

```python
def build_raw_node_features(
    soc_eb,
    soc_pb,
    p_dem,
    acceleration,
):
    soc_values = {
        "energy_battery": soc_eb,
        "power_battery": soc_pb,
        "converter": 0.0,
        "motor": 0.0,
        "vehicle": 0.0,
    }

    continuous = []

    for node_name in GNN_NODE_NAMES:
        static = STATIC_FEATURES[node_name]

        continuous.append([
            soc_values[node_name],
            static["current_discharge_max_a"],
            static["capacity_ah"],
            static["power_discharge_max_w"],
            static["power_charge_max_abs_w"],
            p_dem,
            acceleration,
        ])

    return np.asarray(
        continuous,
        dtype=np.float32,
    )
```

Calcul de la normalisation 

```python
valid_train = (
    train_df["alpha_ems_eb_priority_valid"]
    .astype(bool)
)

train_valid_df = (
    train_df.loc[valid_train]
    .reset_index(drop=True)
)

raw_train_features = []

for row in train_valid_df.itertuples():
    features = build_raw_node_features(
        soc_eb=row.SOC_EB,
        soc_pb=row.SOC_PB,
        p_dem=row.hasPower,
        acceleration=row.hasAcceleration,
    )

    raw_train_features.append(features)

raw_train_features = np.concatenate(
    raw_train_features,
    axis=0,
)

gnn_x_mean = raw_train_features.mean(
    axis=0,
    dtype=np.float64,
).astype(np.float32)

gnn_x_std = raw_train_features.std(
    axis=0,
    dtype=np.float64,
).astype(np.float32)

if np.any(gnn_x_std < 1e-12):
    constant_features = np.array(
        GNN_CONTINUOUS_FEATURE_NAMES
    )[gnn_x_std < 1e-12]

    raise ValueError(
        f"Caractéristiques constantes : "
        f"{constant_features.tolist()}"
    )

np.savez(
    GNN_SCALER_FILE,
    x_mean=gnn_x_mean,
    x_std=gnn_x_std,
    continuous_feature_names=np.array(
        GNN_CONTINUOUS_FEATURE_NAMES
    ),
    node_names=np.array(GNN_NODE_NAMES),
)

print("Moyennes :", gnn_x_mean)
print("Écarts-types :", gnn_x_std)
print("Scalers :", GNN_SCALER_FILE)
```

Encodage des noeuds

```python
NODE_TYPE_ENCODING = np.eye(
    len(GNN_NODE_NAMES),
    dtype=np.float32,
)
```

Construction du graphes

```python
def build_graph(
    soc_eb,
    soc_pb,
    p_dem,
    acceleration,
    alpha_target,
    sample_index,
    time_value,
):
    continuous = build_raw_node_features(
        soc_eb=soc_eb,
        soc_pb=soc_pb,
        p_dem=p_dem,
        acceleration=acceleration,
    )

    continuous = (
        continuous - gnn_x_mean
    ) / gnn_x_std

    node_features = np.concatenate(
        [
            continuous,
            NODE_TYPE_ENCODING,
        ],
        axis=1,
    )

    return Data(
        x=torch.tensor(
            node_features,
            dtype=torch.float32,
        ),
        edge_index=edge_index.clone(),
        y=torch.tensor(
            [alpha_target],
            dtype=torch.float32,
        ),
        sample_index=torch.tensor(
            [sample_index],
            dtype=torch.long,
        ),
        time=torch.tensor(
            [time_value],
            dtype=torch.float32,
        ),
        p_dem=torch.tensor(
            [p_dem],
            dtype=torch.float32,
        ),
    )
```

Construction d'un jeu du graphe 

```python
def build_graph_dataset(data):
    valid = (
        data["alpha_ems_eb_priority_valid"]
        .astype(bool)
    )

    selected = data.loc[valid].copy()

    graphs = []

    for index, row in selected.iterrows():
        graph = build_graph(
            soc_eb=row["SOC_EB"],
            soc_pb=row["SOC_PB"],
            p_dem=row["hasPower"],
            acceleration=row["hasAcceleration"],
            alpha_target=row["alpha_ems_eb_priority"],
            sample_index=index,
            time_value=row[TIME_COL],
        )

        graphs.append(graph)

    return graphs
```

Création du graphe

```python
graphs_train = build_graph_dataset(
    train_df
)

graphs_val = build_graph_dataset(
    val_df
)

graphs_test = build_graph_dataset(
    test_df
)

print("Train      :", len(graphs_train))
print("Validation :", len(graphs_val))
print("Test       :", len(graphs_test))

print(graphs_train[0])
print("x :", graphs_train[0].x.shape)
print("edge_index :", graphs_train[0].edge_index.shape)
print("y :", graphs_train[0].y.shape)
```

Controle du graphe

```python
for split_name, graphs in {
    "train": graphs_train,
    "validation": graphs_val,
    "test": graphs_test,
}.items():
    if len(graphs) == 0:
        raise ValueError(
            f"Aucun graphe dans {split_name}."
        )

    for graph in graphs:
        if graph.x.shape != (
            len(GNN_NODE_NAMES),
            GNN_INPUT_DIM,
        ):
            raise ValueError(
                f"Forme incorrecte dans {split_name}."
            )

        if not torch.isfinite(graph.x).all():
            raise ValueError(
                f"Valeur non finie dans {split_name}."
            )

        if not torch.isfinite(graph.y).all():
            raise ValueError(
                f"Cible non finie dans {split_name}."
            )

        if not (
            ALPHA_MIN
            <= graph.y.item()
            <= ALPHA_MAX
        ):
            raise ValueError(
                f"Cible hors bornes dans {split_name}."
            )

print("Graphes validés.")
```

Sauvegarde

```python
torch.save(
    {
        "train": graphs_train,
        "validation": graphs_val,
        "test": graphs_test,
        "edge_index": edge_index,
        "node_names": GNN_NODE_NAMES,
        "continuous_feature_names": GNN_CONTINUOUS_FEATURE_NAMES,
        "input_dim": GNN_INPUT_DIM,
    },
    GNN_GRAPHS_FILE,
)

print("Graphes sauvegardés :", GNN_GRAPHS_FILE)
print("Scalers sauvegardés :", GNN_SCALER_FILE)
```
