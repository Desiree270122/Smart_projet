```python
# Cellule 3 : regle EB-priority (repli) + diagnostic + filtre de securite
def eb_priority_alpha_single(p_dem, soc_eb):
    if abs(p_dem) <= EPS_POWER_W:
        return 0.5
    if p_dem < P_EB_MIN_W:
        p_eb = P_EB_MIN_W
    elif p_dem < 0:
        p_eb = p_dem
    elif soc_eb <= SOC_EB_MIN:
        p_eb = 0.0
    elif p_dem <= P_EB_MAX_W:
        p_eb = p_dem
    else:
        p_eb = P_EB_MAX_W
    p_pb = p_dem - p_eb
    return float(np.clip(p_pb / p_dem, 0.0, 1.0))


def diagnose_violation(alpha_req, p_dem, soc_eb, soc_pb):
    reasons = []
    if alpha_req < 0 or alpha_req > 1:
        reasons.append('alpha_hors_0_1')
    p_pb = alpha_req * p_dem
    p_eb = p_dem - p_pb
    p_conv = estimate_p_conv(np.array([p_eb]))[0]
    if p_eb < P_EB_MIN_W or p_eb > P_EB_MAX_W: reasons.append('P_EB_hors_limites')
    if p_pb < P_PB_MIN_W or p_pb > P_PB_MAX_W: reasons.append('P_PB_hors_limites')
    if p_conv < P_CONV_MIN_W or p_conv > P_CONV_MAX_W: reasons.append('P_conv_hors_limites')
    if soc_eb <= SOC_EB_MIN + SOC_TOL and p_eb > EPS_POWER_W: reasons.append('EB_protection_SOC_bas')
    if soc_pb <= SOC_PB_MIN + SOC_TOL and p_pb > EPS_POWER_W: reasons.append('PB_protection_SOC_bas')
    i_eb, i_pb = p_eb / V_EB_PACK_NOM, p_pb / V_PB_PACK_NOM
    soc_eb_next = soc_eb - i_eb * DT_OPT_SECONDS / (3600.0 * CAPACITY_EB_AH)
    soc_pb_next = soc_pb - i_pb * DT_OPT_SECONDS / (3600.0 * CAPACITY_PB_AH)
    if soc_eb_next < SOC_EB_MIN - SOC_TOL or soc_eb_next > SOC_EB_MAX + SOC_TOL: reasons.append('SOC_EB_projete_hors_bornes')
    if soc_pb_next < SOC_PB_MIN - SOC_TOL or soc_pb_next > SOC_PB_MAX + SOC_TOL: reasons.append('SOC_PB_projete_hors_bornes')
    return reasons


def safety_filter(alpha_req, p_dem, soc_eb, soc_pb, alpha_prev=None):
    if abs(p_dem) <= EPS_POWER_W:
        return {'alpha_requested': alpha_req, 'alpha_final': alpha_req, 'correction_applied': False,
                'violated_constraint': 'aucune', 'explanation': 'demande nulle, aucune decision necessaire'}

    metrics = candidate_metrics(ALPHA_GRID, p_dem, soc_eb, soc_pb, alpha_prev)
    feasible_idx = np.flatnonzero(metrics['feasible'])

    if len(feasible_idx) == 0:
        alpha_final = eb_priority_alpha_single(p_dem, soc_eb)
        return {'alpha_requested': alpha_req, 'alpha_final': alpha_final, 'correction_applied': True,
                'violated_constraint': 'aucune_solution_faisable',
                'explanation': 'Aucune repartition electrique faisable (probable freinage mecanique requis) ; repli sur la regle EB-priority.'}

    feasible_alphas = ALPHA_GRID[feasible_idx]
    nearest = feasible_idx[np.argmin(np.abs(feasible_alphas - alpha_req))]
    alpha_final = float(ALPHA_GRID[nearest])
    corrected = not np.isclose(alpha_final, alpha_req, atol=ALPHA_GRID_STEP * 1.1)
    reasons = diagnose_violation(alpha_req, p_dem, soc_eb, soc_pb) if corrected else []

    return {
        'alpha_requested': alpha_req, 'alpha_final': alpha_final, 'correction_applied': corrected,
        'violated_constraint': ','.join(reasons) if reasons else 'aucune',
        'explanation': 'aucune correction necessaire' if not corrected else
                       f"Projection vers la decision faisable la plus proche (causes : {', '.join(reasons)})",
    }
```

```python
# Cellule 4 : application sequentielle du filtre sur alpha_fusion (test)
journal_rows = []
alpha_prev_filter = None

for i in range(len(test_sub)):
    p_dem = p_test[i]; soc_eb = soceb_test[i]; soc_pb = socpb_test[i]
    result = safety_filter(alpha_fusion_test[i], p_dem, soc_eb, soc_pb, alpha_prev_filter)

    p_pb_final = result['alpha_final'] * p_dem
    p_eb_final = p_dem - p_pb_final

    journal_rows.append({
        'time': test_sub[TIME_COL].iloc[i],
        'alpha_ems_fuzzy_logic': fuzzy_test[i], 'alpha_ems_mlp_neurosymbolic': alpha_mlp_ns_test[i],
        'alpha_requested': result['alpha_requested'], 'alpha_final': result['alpha_final'],
        'P_EB': p_eb_final, 'P_PB': p_pb_final,
        'correction_applied': result['correction_applied'],
        'violated_constraint': result['violated_constraint'],
        'explanation': result['explanation'],
    })

    if abs(p_dem) > EPS_POWER_W:
        alpha_prev_filter = result['alpha_final']

journal_df = pd.DataFrame(journal_rows)
print('Corrections appliquees :', int(journal_df['correction_applied'].sum()), '/', len(journal_df))
display(journal_df['violated_constraint'].value_counts())
```
