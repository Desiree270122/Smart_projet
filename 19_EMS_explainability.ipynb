```python
# Cellule 5 : TOUTES les classes des modeles boite noire + chargeurs (autonome, rien ne manque)

class MLPSimpleModel(nn.Module):
    """MLP simple -- attribut 'net', Sigmoid en module (additivite SHAP)."""
    def __init__(self, n_in, h1, h2, dropout):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n_in, h1), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(h1, h2), nn.ReLU(), nn.Linear(h2, 1),
            nn.Sigmoid(),
        )
    def forward(self, x):
        return self.net(x).squeeze(-1)


class DirectAlphaMLP(nn.Module):
    """Variantes d'ablation -- attribut 'network', Sigmoid en module."""
    def __init__(self, n_in, h1, h2, dropout):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(n_in, h1), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(h1, h2), nn.ReLU(), nn.Linear(h2, 1),
            nn.Sigmoid(),
        )
    def forward(self, x):
        return self.network(x).squeeze(-1)


class MLPNeuroSymbolique(nn.Module):
    """MLP-NS deploye -- residuel borne, retourne alpha, delta_alpha, ET alpha_unclamped (avant clip)."""
    def __init__(self, n_in, hidden1, hidden2, dropout, max_delta):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n_in, hidden1), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(hidden1, hidden2), nn.ReLU(),
            nn.Linear(hidden2, 1),
        )
        self.max_delta = max_delta
        self.tanh = nn.Tanh()
        self.clip = nn.Hardtanh(0.0, 1.0)

    def forward(self, x, alpha_ems_fuzzy_logic):
        z = self.net(x).squeeze(-1)
        delta_alpha = self.max_delta * self.tanh(z)
        alpha_unclamped = alpha_ems_fuzzy_logic + delta_alpha
        alpha = self.clip(alpha_unclamped)
        return alpha, delta_alpha, alpha_unclamped


class LSTMSeul(nn.Module):
    """LSTM seul -- attribut 'fc'."""
    def __init__(self, n_features, hidden_size, num_layers, dropout):
        super().__init__()
        self.lstm = nn.LSTM(n_features, hidden_size, num_layers, batch_first=True,
                             dropout=dropout if num_layers > 1 else 0.0)
        self.fc = nn.Sequential(nn.Linear(hidden_size, 32), nn.ReLU(), nn.Linear(32, 3))

    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])


class LSTMNeuroSymbolique(nn.Module):
    """LSTM neurosymbolique -- attribut 'head' (distinct du LSTM seul)."""
    def __init__(self, n_features, hidden_size, num_layers, dropout):
        super().__init__()
        self.lstm = nn.LSTM(n_features, hidden_size, num_layers, batch_first=True,
                             dropout=dropout if num_layers > 1 else 0.0)
        self.head = nn.Sequential(nn.Linear(hidden_size, 32), nn.ReLU(), nn.Linear(32, 3))

    def forward(self, x):
        out, _ = self.lstm(x)
        return self.head(out[:, -1, :])


class MLPWrapper(nn.Module):
    """Wrapper SHAP generique pour les modeles a une seule entree x -> scalaire."""
    def __init__(self, base_model):
        super().__init__()
        self.base_model = base_model
    def forward(self, x):
        return self.base_model(x).reshape(x.shape[0], 1)


ABLATION_HIDDEN_1 = int(globals().get("MLP_SIMPLE_HIDDEN_1", 64))
ABLATION_HIDDEN_2 = int(globals().get("MLP_SIMPLE_HIDDEN_2", 32))
ABLATION_DROPOUT = float(globals().get("MLP_SIMPLE_DROPOUT", 0.10))


def load_mlp_simple():
    scalers = np.load(MODELS_DIR / 'EMS_MLP_scalers.npz', allow_pickle=True)
    cols = list(scalers['input_cols'])
    model = MLPSimpleModel(len(cols), MLP_SIMPLE_HIDDEN_1, MLP_SIMPLE_HIDDEN_2, MLP_SIMPLE_DROPOUT).to(DEVICE)
    try:
        sd = torch.load(CHECKPOINTS_DIR / 'EMS_MLP.pt', map_location=DEVICE, weights_only=True)
    except TypeError:
        sd = torch.load(CHECKPOINTS_DIR / 'EMS_MLP.pt', map_location=DEVICE)
    model.load_state_dict(sd)
    model.eval()
    return model, scalers['x_mean'], scalers['x_std'], cols


def load_ablation_best():
    import json
    sel_file = MODELS_DIR / 'ablation_selected_model.json'
    selection = json.load(open(sel_file, encoding='utf-8'))
    scaler_file = Path(selection['scaler'])
    scalers = np.load(scaler_file, allow_pickle=True)
    cols = list(scalers['input_cols'])
    model = DirectAlphaMLP(len(cols), ABLATION_HIDDEN_1, ABLATION_HIDDEN_2, ABLATION_DROPOUT).to(DEVICE)
    try:
        sd = torch.load(Path(selection['checkpoint']), map_location=DEVICE, weights_only=True)
    except TypeError:
        sd = torch.load(Path(selection['checkpoint']), map_location=DEVICE)
    model.load_state_dict(sd)
    model.eval()
    return model, scalers['x_mean'], scalers['x_std'], cols, selection['best_variant']


def load_mlp_ns_deployed():
    """MLP-NS deploye -- cherche le scaler/checkpoint parmi plusieurs noms possibles."""
    scaler_candidates = [
        'EMS_MLP_neurosymbolic_scalers.npz', 'mlp_neurosymbolique_scalers.npz',
        'mlp_ns_scalers.npz', 'EMS_MLP_neurosymbolic_scaler.npz',
    ]
    checkpoint_candidates = [
        'EMS_MLP_neurosymbolic.pt', 'meilleur_mlp_neurosymbolique.pt', 'mlp_neurosymbolique.pt',
    ]
    scaler_file = next((MODELS_DIR / n for n in scaler_candidates if (MODELS_DIR / n).exists()), None)
    checkpoint_file = next((CHECKPOINTS_DIR / n for n in checkpoint_candidates if (CHECKPOINTS_DIR / n).exists()), None)
    if scaler_file is None or checkpoint_file is None:
        raise FileNotFoundError(
            f"Scaler trouve : {scaler_file} | Checkpoint trouve : {checkpoint_file}\n"
            f"Scalers disponibles : {sorted(p.name for p in MODELS_DIR.glob('*.npz'))}\n"
            f"Checkpoints disponibles : {sorted(p.name for p in CHECKPOINTS_DIR.glob('*.pt'))}"
        )
    print(f"Scaler utilise : {scaler_file.name} | Checkpoint utilise : {checkpoint_file.name}")
    scalers = np.load(scaler_file, allow_pickle=True)
    cols = list(scalers['input_cols'])
    model = MLPNeuroSymbolique(len(cols), MLP_NS_HIDDEN_1, MLP_NS_HIDDEN_2, MLP_NS_DROPOUT, MAX_DELTA_ALPHA).to(DEVICE)
    try:
        sd = torch.load(checkpoint_file, map_location=DEVICE, weights_only=True)
    except TypeError:
        sd = torch.load(checkpoint_file, map_location=DEVICE)
    model.load_state_dict(sd)
    model.eval()
    return model, scalers['x_mean'], scalers['x_std'], cols


def load_lstm_seul():
    scalers = np.load(MODELS_DIR / 'EMS_LSTM_scalers.npz', allow_pickle=True)
    cols = list(scalers['input_cols'])
    model = LSTMSeul(len(cols), LSTM_HIDDEN_SIZE, LSTM_NUM_LAYERS, LSTM_DROPOUT).to(DEVICE)
    try:
        sd = torch.load(CHECKPOINTS_DIR / 'EMS_LSTM.pt', map_location=DEVICE, weights_only=True)
    except TypeError:
        sd = torch.load(CHECKPOINTS_DIR / 'EMS_LSTM.pt', map_location=DEVICE)
    model.load_state_dict(sd)
    model.eval()
    return model, scalers['x_mean'], scalers['x_std'], scalers['y_mean'], scalers['y_std'], cols


def load_lstm_ns():
    scalers = np.load(MODELS_DIR / 'EMS_LSTM_neurosymbolic_scalers.npz', allow_pickle=True)
    cols = list(scalers['input_cols'])
    model = LSTMNeuroSymbolique(len(cols), LSTM_NS_HIDDEN_SIZE, LSTM_NS_NUM_LAYERS, LSTM_NS_DROPOUT).to(DEVICE)
    try:
        sd = torch.load(CHECKPOINTS_DIR / 'EMS_LSTM_neurosymbolic.pt', map_location=DEVICE, weights_only=True)
    except TypeError:
        sd = torch.load(CHECKPOINTS_DIR / 'EMS_LSTM_neurosymbolic.pt', map_location=DEVICE)
    model.load_state_dict(sd)
    model.eval()
    return model, scalers['x_mean'], scalers['x_std'], scalers['y_mean'], scalers['y_std'], cols
```

```python
# Cellule 6 : SHAP pour le MLP simple
import shap

mlp_model, mlp_x_mean, mlp_x_std, mlp_cols = load_mlp_simple()
mlp_model.eval()

col_values = {'SOC_EB': soc_eb_arr, 'SOC_PB': soc_pb_arr, 'hasPower': pdem_arr,
              'speed': test_full['speed'].to_numpy(), 'hasAcceleration': accel_arr}
X_mlp = np.stack([col_values[c] for c in mlp_cols], axis=1).astype(np.float32)
X_mlp_n = (X_mlp - mlp_x_mean) / mlp_x_std

background_idx = RNG.choice(len(X_mlp_n), size=min(200, len(X_mlp_n)), replace=False)
sample_idx = RNG.choice(len(X_mlp_n), size=min(500, len(X_mlp_n)), replace=False)

background = torch.from_numpy(X_mlp_n[background_idx]).float().to(DEVICE)
sample = torch.from_numpy(X_mlp_n[sample_idx]).float().to(DEVICE)

explainer_mlp = shap.DeepExplainer(MLPWrapper(mlp_model), background)
shap_values_mlp = explainer_mlp.shap_values(sample)
if isinstance(shap_values_mlp, list):
    shap_values_mlp = shap_values_mlp[0]
shap_values_mlp = np.asarray(shap_values_mlp).squeeze()
if shap_values_mlp.shape != (len(sample), len(mlp_cols)):
    raise ValueError(f"Forme SHAP inattendue : {shap_values_mlp.shape}, attendu {(len(sample), len(mlp_cols))}")

shap_summary_mlp = pd.DataFrame({
    'variable': mlp_cols,
    'contribution_moyenne_absolue': np.abs(shap_values_mlp).mean(axis=0),
    'contribution_moyenne_signee': shap_values_mlp.mean(axis=0),
}).sort_values('contribution_moyenne_absolue', ascending=False)
display(shap_summary_mlp)
shap_summary_mlp.to_csv(TABLES_DIR / 'results_explicabilite_shap_mlp.csv', index=False)
```

```python
# Cellule 8 : SHAP pour le MLP-NS deploye (alpha_unclamped, alignement NaN, unpacking a 3 valeurs)
mlp_ns_model, mns_x_mean, mns_x_std, mns_cols = load_mlp_ns_deployed()
mlp_ns_model.eval()

missing_cols_ns = [c for c in mns_cols if c not in test_full.columns and c != 'alpha_ems_fuzzy_logic']
if missing_cols_ns:
    raise ValueError(f"Colonnes manquantes pour le modele neurosymbolique deploye : {missing_cols_ns}")

ALPHA_FUZZY_IN_X = 'alpha_ems_fuzzy_logic' in mns_cols
print(f"alpha_ems_fuzzy_logic deja present dans les colonnes d'entree x : {ALPHA_FUZZY_IN_X}")

X_ns_dict = {c: (alpha_fz if c == 'alpha_ems_fuzzy_logic' else test_full[c].to_numpy()) for c in mns_cols}
X_ns = np.stack([X_ns_dict[c] for c in mns_cols], axis=1).astype(np.float32)

valid_rows_ns = ~np.isnan(X_ns).any(axis=1)
n_dropped_ns = int((~valid_rows_ns).sum())
print(f"Lignes avec au moins un NaN exclues avant SHAP : {n_dropped_ns} / {len(X_ns)}")

X_ns_valid = X_ns[valid_rows_ns]
alpha_fz_valid = alpha_fz[valid_rows_ns]
X_ns_n = (X_ns_valid - mns_x_mean) / mns_x_std
if not np.isfinite(X_ns_n).all():
    raise ValueError("Valeurs non finies persistantes apres filtrage des NaN.")

background_idx_ns = RNG.choice(len(X_ns_n), size=min(200, len(X_ns_n)), replace=False)
sample_idx_ns = RNG.choice(len(X_ns_n), size=min(500, len(X_ns_n)), replace=False)

if ALPHA_FUZZY_IN_X:
    alpha_idx_in_x = mns_cols.index('alpha_ems_fuzzy_logic')
    Z_background = X_ns_n[background_idx_ns]
    Z_sample = X_ns_n[sample_idx_ns]
    n_features_z = Z_background.shape[1]

    class MLPNSWrapperShared(nn.Module):
        def __init__(self, base_model, alpha_idx, x_mean_t, x_std_t):
            super().__init__()
            self.base_model = base_model
            self.alpha_idx = alpha_idx
            self.x_mean_t = x_mean_t
            self.x_std_t = x_std_t
        def forward(self, z):
            alpha_fuzzy_denorm = z[:, self.alpha_idx] * self.x_std_t[self.alpha_idx] + self.x_mean_t[self.alpha_idx]
            alpha, delta_alpha, alpha_unclamped = self.base_model(z, alpha_fuzzy_denorm)
            return alpha_unclamped.reshape(z.shape[0], 1), delta_alpha.reshape(z.shape[0], 1)

    x_mean_t = torch.from_numpy(mns_x_mean).float().to(DEVICE)
    x_std_t = torch.from_numpy(mns_x_std).float().to(DEVICE)
    wrapper_base = MLPNSWrapperShared(mlp_ns_model, alpha_idx_in_x, x_mean_t, x_std_t)
else:
    alpha_fz_n = (alpha_fz_valid - alpha_fz_valid.mean()) / (alpha_fz_valid.std() + 1e-9)
    Z_full = np.column_stack([X_ns_n, alpha_fz_n]).astype(np.float32)
    Z_background = Z_full[background_idx_ns]
    Z_sample = Z_full[sample_idx_ns]
    n_features_z = Z_background.shape[1]
    alpha_fz_mean_raw, alpha_fz_std_raw = float(alpha_fz_valid.mean()), float(alpha_fz_valid.std())

    class MLPNSWrapperConcat(nn.Module):
        def __init__(self, base_model, mean_raw, std_raw):
            super().__init__()
            self.base_model = base_model
            self.mean_raw = mean_raw
            self.std_raw = std_raw
        def forward(self, z):
            x = z[:, :-1]
            alpha_fuzzy_denorm = z[:, -1] * self.std_raw + self.mean_raw
            alpha, delta_alpha, alpha_unclamped = self.base_model(x, alpha_fuzzy_denorm)
            return alpha_unclamped.reshape(z.shape[0], 1), delta_alpha.reshape(z.shape[0], 1)

    wrapper_base = MLPNSWrapperConcat(mlp_ns_model, alpha_fz_mean_raw, alpha_fz_std_raw)


background_ns = torch.from_numpy(Z_background).float().to(DEVICE)
sample_ns = torch.from_numpy(Z_sample).float().to(DEVICE)
feature_names_z = list(mns_cols) if ALPHA_FUZZY_IN_X else list(mns_cols) + ['alpha_ems_fuzzy_logic']

class FinalOutputWrapper(nn.Module):
    def __init__(self, wb): super().__init__(); self.wb = wb
    def forward(self, z): return self.wb(z)[0]

explainer_final = shap.DeepExplainer(FinalOutputWrapper(wrapper_base), background_ns)
shap_final = explainer_final.shap_values(sample_ns)
if isinstance(shap_final, list): shap_final = shap_final[0]
shap_final = np.asarray(shap_final).squeeze()

expected_shape = (len(sample_ns), n_features_z)
if shap_final.shape != expected_shape:
    raise ValueError(f"Forme SHAP (alpha_final) inattendue : {shap_final.shape}, attendu {expected_shape}")

shap_summary_final = pd.DataFrame({
    'variable': feature_names_z,
    'contribution_moyenne_absolue': np.abs(shap_final).mean(axis=0),
    'contribution_moyenne_signee': shap_final.mean(axis=0),
}).sort_values('contribution_moyenne_absolue', ascending=False)
display(shap_summary_final)
shap_summary_final.to_csv(TABLES_DIR / 'results_explicabilite_shap_mlp_neurosymbolic_alpha_avant_ecretage.csv', index=False)

class DeltaAlphaWrapper(nn.Module):
    def __init__(self, wb): super().__init__(); self.wb = wb
    def forward(self, z): return self.wb(z)[1]

explainer_delta = shap.DeepExplainer(DeltaAlphaWrapper(wrapper_base), background_ns)
shap_delta = explainer_delta.shap_values(sample_ns)
if isinstance(shap_delta, list): shap_delta = shap_delta[0]
shap_delta = np.asarray(shap_delta).squeeze()
if shap_delta.shape != expected_shape:
    raise ValueError(f"Forme SHAP (delta_alpha) inattendue : {shap_delta.shape}, attendu {expected_shape}")

shap_summary_delta = pd.DataFrame({
    'variable': feature_names_z,
    'contribution_moyenne_absolue': np.abs(shap_delta).mean(axis=0),
    'contribution_moyenne_signee': shap_delta.mean(axis=0),
}).sort_values('contribution_moyenne_absolue', ascending=False)
display(shap_summary_delta)
shap_summary_delta.to_csv(TABLES_DIR / 'results_explicabilite_shap_mlp_neurosymbolic_delta_alpha.csv', index=False)

print(f"\nalpha_fuzzy moyen (composante symbolique) : {alpha_fz_valid[sample_idx_ns].mean():.4f}")
```

```python
# Cellule 9 : Integrated Gradients (Captum) pour LSTM seul et LSTM-NS -- cuDNN desactive temporairement
from captum.attr import IntegratedGradients

def explain_lstm_ig(model, x_mean, x_std, input_cols, df, window=WINDOW, horizon=HORIZON, n_samples=100):
    model.eval()
    X = df[input_cols].to_numpy(dtype=np.float32)
    Xn = (X - x_mean) / x_std
    n_valid = len(Xn) - window - horizon + 1
    n = min(n_samples, n_valid)
    sample_anchors = RNG.choice(n_valid, size=n, replace=False)

    windows = np.stack([Xn[a:a + window] for a in sample_anchors])
    windows_t = torch.from_numpy(windows).float().to(DEVICE).requires_grad_(True)
    baseline_t = torch.zeros_like(windows_t)

    ig = IntegratedGradients(model)
    output_names = ['Pdem', 'delta_SOC_EB', 'delta_SOC_PB']
    attributions_by_output = {}

    with torch.backends.cudnn.flags(enabled=False):
        for out_idx, out_name in enumerate(output_names):
            attr, delta = ig.attribute(windows_t, baselines=baseline_t, target=out_idx, n_steps=50,
                                         return_convergence_delta=True)
            max_delta = float(delta.abs().max())
            if max_delta > 0.05:
                print(f"ATTENTION [{out_name}] : delta de convergence max = {max_delta:.4f}")
            attributions_by_output[out_name] = attr.detach().cpu().numpy()

    return attributions_by_output


def summarize_ig(attributions_by_output, cols, label):
    rows, time_curves = [], {}
    for out_name, attr in attributions_by_output.items():
        var_importance = np.abs(attr).sum(axis=1).mean(axis=0)
        time_curves[out_name] = np.abs(attr).sum(axis=2).mean(axis=0)
        for i, c in enumerate(cols):
            rows.append({'modele': label, 'sortie': out_name, 'variable': c, 'importance_moyenne': float(var_importance[i])})
    return pd.DataFrame(rows), time_curves


lstm_seul_model, ls_x_mean, ls_x_std, ls_y_mean, ls_y_std, ls_cols = load_lstm_seul()
attr_lstm_seul = explain_lstm_ig(lstm_seul_model, ls_x_mean, ls_x_std, ls_cols, test_full)

lstm_ns_model, ln_x_mean, ln_x_std, ln_y_mean, ln_y_std, ln_cols = load_lstm_ns()
attr_lstm_ns = explain_lstm_ig(lstm_ns_model, ln_x_mean, ln_x_std, ln_cols, test_full)

ig_summary_seul, time_curves_seul = summarize_ig(attr_lstm_seul, ls_cols, 'EMS_LSTM')
ig_summary_ns, time_curves_ns = summarize_ig(attr_lstm_ns, ln_cols, 'EMS_LSTM_neurosymbolic')

ig_summary = pd.concat([ig_summary_seul, ig_summary_ns], ignore_index=True)
display(ig_summary.sort_values(['modele', 'sortie', 'importance_moyenne'], ascending=[True, True, False]))
ig_summary.to_csv(TABLES_DIR / 'results_explicabilite_integrated_gradients.csv', index=False)

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=False)
for ax, out_name in zip(axes, ['Pdem', 'delta_SOC_EB', 'delta_SOC_PB']):
    ax.plot(range(-WINDOW + 1, 1), time_curves_seul[out_name], label='EMS_LSTM', marker='o')
    ax.plot(range(-WINDOW + 1, 1), time_curves_ns[out_name], label='EMS_LSTM_neurosymbolic', marker='o')
    ax.set_title(out_name)
    ax.set_xlabel('Pas dans la fenetre (0 = present)')
axes[0].set_ylabel('Importance moyenne (|attribution|)')
axes[0].legend()
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'explicabilite_ig_importance_temporelle.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cellule 12 : generation automatique d'explications naturelles pour le MLP neurosymbolique

import numpy as np
import pandas as pd
import torch

print(mns_cols)
print('alpha_ems_fuzzy_logic dans mns_cols :', 'alpha_ems_fuzzy_logic' in mns_cols)


# 1. Noms comprehensibles pour les variables et les regles

FEATURE_LABELS = {
    "hasPower": "la puissance demandée",
    "SOC_EB": "le niveau de charge de la batterie d'énergie",
    "SOC_PB": "le niveau de charge de la batterie de puissance",
    "speed": "la vitesse du véhicule",
    "hasAcceleration": "l'accélération",
    "hasTotalForce": "la force totale appliquée au véhicule",
    "I_EB": "le courant de la batterie d'énergie",
    "I_PB": "le courant de la batterie de puissance",
    "alpha_ems_fuzzy_logic": "la décision initiale de la logique floue",
    "Pdem_pred": "la puissance future prédite",
    "delta_SOC_EB_pred": "la variation future du SOC de la batterie d'énergie",
    "delta_SOC_PB_pred": "la variation future du SOC de la batterie de puissance",
    "converter_risk": "le risque de surcharge du convertisseur",
    "high_power_demand": "une forte demande de puissance",
    "regenerative_braking": "le freinage régénératif",
    "zero_power_demand": "une demande de puissance quasi nulle",
}

RULE_LABELS = {
    "R1_PB_faible_traction": "la protection de la batterie de puissance lorsque son niveau de charge est faible",
    "R2_EB_faible_protection": "la préservation de la batterie d'énergie lorsque son niveau de charge est faible",
    "R3_forte_traction_PB_dispo": "l'utilisation de la batterie de puissance lors d'une forte demande de traction",
    "R4_demande_nulle_stable": "le maintien d'une répartition stable lorsque la demande de puissance est faible",
    "R5_recharge_vers_PB": "l'orientation de la récupération d'énergie vers la batterie de puissance",
    "R5_regenerative_braking": "l'orientation de la récupération d'énergie vers la batterie de puissance",
    "R5b_PB_high_recharge": "l'orientation prioritaire de la recharge vers la batterie de puissance lorsqu'elle est déjà bien chargée",
    "R7_deux_SOC_bas": "la protection simultanée des deux batteries lorsque leurs niveaux de charge sont faibles",
    "aucune_regle_dominante": "aucune règle floue particulière",
}


def nom_humain_variable(nom):
    return FEATURE_LABELS.get(nom, nom.replace("_", " "))


def nom_humain_regle(nom):
    return RULE_LABELS.get(nom, nom.replace("_", " "))


def joindre_liste(elements):
    elements = list(elements)
    if len(elements) == 0:
        return "aucun facteur clairement identifié"
    if len(elements) == 1:
        return elements[0]
    if len(elements) == 2:
        return f"{elements[0]} et {elements[1]}"
    return ", ".join(elements[:-1]) + f" et {elements[-1]}"


# 2. Construction securisee des entrees

def construire_x_full(data, alpha_fuzzy_arr, cols):
    colonnes_manquantes = [c for c in cols if c != "alpha_ems_fuzzy_logic" and c not in data.columns]
    if colonnes_manquantes:
        raise ValueError(f"Colonnes manquantes pour le MLP neurosymbolique : {colonnes_manquantes}")

    X_dict = {}
    for c in cols:
        X_dict[c] = alpha_fuzzy_arr if c == "alpha_ems_fuzzy_logic" else data[c].to_numpy()

    return np.stack([X_dict[c] for c in cols], axis=1).astype(np.float32)


X_ns_full = construire_x_full(test_full, alpha_fz, mns_cols)

valid_full = np.isfinite(X_ns_full).all(axis=1) & np.isfinite(alpha_fz)
n_invalid = int((~valid_full).sum())

print(f"Lignes ignorées, car les entrées du MLP neurosymbolique sont incomplètes : "
      f"{n_invalid} / {len(test_full)} ({100 * n_invalid / len(test_full):.1f} %)")


# 3. Inference uniquement sur les lignes valides

n_total = len(test_full)

# alpha_model_arr : sortie BRUTE du reseau (apres Hardtanh), PAS la decision physique finale --
# le vrai filtre de securite (safety_filter) n'est pas encore appele a ce stade.
alpha_model_arr = np.full(n_total, np.nan, dtype=np.float32)
delta_alpha_arr = np.full(n_total, np.nan, dtype=np.float32)

valid_indices = np.flatnonzero(valid_full)
if len(valid_indices) == 0:
    raise ValueError("Aucune ligne valide n'est disponible pour l'inférence.")

X_valid = X_ns_full[valid_indices]

# Protection division par zero si une colonne est constante dans le train
mns_x_std_safe = np.where(np.abs(mns_x_std) < 1e-12, 1.0, mns_x_std)
X_valid_n = (X_valid - mns_x_mean) / mns_x_std_safe

alpha_fuzzy_valid = alpha_fz[valid_indices].astype(np.float32)

with torch.no_grad():
    alpha_valid, delta_valid, _ = mlp_ns_model(
        torch.from_numpy(X_valid_n).float().to(DEVICE),
        torch.from_numpy(alpha_fuzzy_valid).float().to(DEVICE),
    )

alpha_model_arr[valid_indices] = alpha_valid.detach().cpu().numpy().reshape(-1)
delta_alpha_arr[valid_indices] = delta_valid.detach().cpu().numpy().reshape(-1)


# 4. Verification physique instantanee -- candidate_metrics exige des scalaires, donc boucle
# NOTE : ceci verifie la FAISABILITE de alpha_model_arr, ce n'est pas un appel au vrai
# safety_filter. Si feasible=False, cela signifie que cette valeur DEVRAIT etre corrigee,
# pas qu'elle l'a deja ete.

feasible_full = np.zeros(n_total, dtype=bool)

for i in valid_indices:
    metrics_i = candidate_metrics(
        float(alpha_model_arr[i]), float(pdem_arr[i]), float(soc_eb_arr[i]), float(soc_pb_arr[i]), None,
    )
    feasible_full[i] = bool(metrics_i['feasible'])


# 5. Facteurs SHAP (importance GLOBALE -- explicitement etiquetee comme telle)

top_facteurs_globaux_bruts = shap_summary_final.head(3)["variable"].tolist()
top_facteurs_globaux = [nom_humain_variable(c) for c in top_facteurs_globaux_bruts]

shap_local_full = globals().get("shap_values_ns_full", None)


def obtenir_facteurs_shap(i, top_k=3):
    if (
        shap_local_full is not None
        and np.asarray(shap_local_full).shape == (len(test_full), len(mns_cols))
        and np.isfinite(shap_local_full[i]).all()
    ):
        valeurs = np.asarray(shap_local_full[i])
        indices = np.argsort(np.abs(valeurs))[::-1][:top_k]
        facteurs = [nom_humain_variable(mns_cols[j]) for j in indices]
        return facteurs, "locale"

    return top_facteurs_globaux, "globale"


# 6. Fonctions de redaction naturelle

EPS_TEXTE_W = float(globals().get("EPS_POWER_W", 100.0))


def decrire_repartition(alpha, pdem):
    p_pb_kw = alpha * pdem / 1000
    p_eb_kw = (1 - alpha) * pdem / 1000

    if pdem > EPS_TEXTE_W:
        return (f"Pour répondre à une demande de {pdem / 1000:.2f} kW, "
                f"la batterie de puissance fournit {p_pb_kw:.2f} kW "
                f"et la batterie d'énergie fournit {p_eb_kw:.2f} kW.")

    if pdem < -EPS_TEXTE_W:
        return (f"Pendant la récupération d'énergie, une puissance de "
                f"{abs(pdem) / 1000:.2f} kW est répartie entre les batteries. "
                f"La batterie de puissance absorbe {abs(p_pb_kw):.2f} kW "
                f"et la batterie d'énergie absorbe {abs(p_eb_kw):.2f} kW.")

    return "La demande de puissance est presque nulle ; aucune sollicitation énergétique importante n'est requise."


def decrire_correction(delta_alpha):
    correction_points = 100 * delta_alpha

    if abs(correction_points) < 0.5:
        return "Le réseau neuronal a conservé presque entièrement la décision proposée par la logique floue."

    if correction_points > 0:
        return (f"Le réseau neuronal a ensuite augmenté de {correction_points:.1f} points de pourcentage "
                f"la part attribuée à la batterie de puissance.")

    return (f"Le réseau neuronal a ensuite diminué de {abs(correction_points):.1f} points de pourcentage "
            f"la part attribuée à la batterie de puissance.")


def decrire_contraintes(i):
    """Le filtre de securite n'est pas reellement appele -- formulation au conditionnel, jamais au passe."""
    if feasible_full[i]:
        return "À cet instant, la décision proposée par le modèle respecte toutes les contraintes physiques vérifiées."

    return ("La décision proposée par le modèle viole au moins une contrainte physique vérifiée "
            "et devrait être corrigée par le filtre de sécurité avant d'être appliquée.")


# 7. Generateur de texte par ligne

def generer_explication_row(i):
    if not valid_full[i]:
        return ("L'explication ne peut pas être calculée à cet instant, car certaines données requises "
                "par le modèle ne sont pas disponibles.")

    row = test_full.iloc[i]
    temps = float(row[TIME_COL])
    pdem = float(row["hasPower"])

    alpha_fuzzy_val = float(alpha_fz[i])
    alpha_model_val = float(alpha_model_arr[i])
    delta_alpha_val = float(delta_alpha_arr[i])

    regle_brute = rule_trace_df.iloc[i]["regle_dominante"]
    regle_humaine = nom_humain_regle(regle_brute)
    poids_regle = float(rule_trace_df.iloc[i]["poids_regle_dominante"])

    facteurs, portee_shap = obtenir_facteurs_shap(i)

    texte = (
        f"À l'instant {temps:.0f} s, le MLP neurosymbolique propose d'attribuer {100 * alpha_model_val:.1f} % "
        f"de la puissance à la batterie de puissance et {100 * (1 - alpha_model_val):.1f} % "
        f"à la batterie d'énergie. "
    )
    texte += decrire_repartition(alpha_model_val, pdem)
    texte += (
        f" La logique floue proposait initialement {100 * alpha_fuzzy_val:.1f} % pour la batterie "
        f"de puissance. La règle la plus active concernait {regle_humaine}, avec un degré d'activation "
        f"de {poids_regle:.2f}. "
    )
    texte += decrire_correction(delta_alpha_val)

    if portee_shap == "locale":
        texte += f" Pour cette décision particulière, SHAP indique que les facteurs les plus influents sont {joindre_liste(facteurs)}. "
    else:
        texte += (f" À l'échelle de l'ensemble du jeu de test, l'analyse SHAP classe parmi les variables "
                  f"les plus influentes {joindre_liste(facteurs)}. Cette importance est globale et ne "
                  f"constitue pas une explication spécifique à cet instant. ")

    texte += decrire_contraintes(i)
    return texte


# 8. Tableau final

explanations_df = pd.DataFrame({
    TIME_COL: test_full[TIME_COL],
    "alpha_ems_fuzzy_logic": alpha_fz,
    "alpha_ems_mlp_neurosymbolic_propose": alpha_model_arr,  # renomme : ce n'est pas encore la decision finale
    "delta_alpha": delta_alpha_arr,
    "regle_dominante": rule_trace_df["regle_dominante"].to_numpy(),
    "poids_regle_dominante": rule_trace_df["poids_regle_dominante"].to_numpy(),
    "respecte_contraintes_physiques": feasible_full,
    "donnees_disponibles": valid_full,
})

explanations_df["explication_texte"] = [generer_explication_row(i) for i in range(len(test_full))]
explanations_df["methode_explication"] = "Règles floues + SHAP + verification physique (filtre non encore applique)"
explanations_df["portee_shap"] = [
    obtenir_facteurs_shap(i)[1] if valid_full[i] else "indisponible"
    for i in range(len(test_full))
]

output_path = TABLES_DIR / "results_explicabilite_texte_genere.csv"
explanations_df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"\n{len(explanations_df)} explications générées et sauvegardées dans : {output_path}")
print(f"Lignes avec données disponibles : {valid_full.sum()} / {len(test_full)} ({100 * valid_full.mean():.1f} %)")

indices_exemples = [min(10, len(test_full) - 1), len(test_full) // 2, max(0, len(test_full) - 10)]
for i in indices_exemples:
    print("\n" + "=" * 80)
    print(explanations_df.iloc[i]["explication_texte"])
```

```python
# Cellule 13 : alpha pour toutes les strategies disponibles (GNN exclu -- non implemente)
alpha_eb_priority_full = test_full['alpha_ems_eb_priority'].to_numpy()
alpha_fuzzy_full = alpha_fz

mlp_model_full, mlp_x_mean_full, mlp_x_std_full, mlp_cols_full = load_mlp_simple()
mlp_x_std_full_safe = np.where(np.abs(mlp_x_std_full) < 1e-12, 1.0, mlp_x_std_full)
col_values_full = {'SOC_EB': soc_eb_arr, 'SOC_PB': soc_pb_arr, 'hasPower': pdem_arr,
                    'speed': test_full['speed'].to_numpy(), 'hasAcceleration': accel_arr}
X_mlp_full = np.stack([col_values_full[c] for c in mlp_cols_full], axis=1).astype(np.float32)
X_mlp_full_n = (X_mlp_full - mlp_x_mean_full) / mlp_x_std_full_safe
with torch.no_grad():
    alpha_mlp_simple_full = mlp_model_full(torch.from_numpy(X_mlp_full_n).float().to(DEVICE)).cpu().numpy()

ablation_model_full, abl_x_mean_full, abl_x_std_full, abl_cols_full, variant_name_full = load_ablation_best()
abl_x_std_full_safe = np.where(np.abs(abl_x_std_full) < 1e-12, 1.0, abl_x_std_full)
X_abl_full_dict = {c: (alpha_fz if c == 'alpha_ems_fuzzy_logic' else test_full[c].to_numpy()) for c in abl_cols_full}
X_abl_full = np.stack([X_abl_full_dict[c] for c in abl_cols_full], axis=1).astype(np.float32)
valid_abl_full = ~np.isnan(X_abl_full).any(axis=1)
X_abl_full_n = np.full_like(X_abl_full, np.nan)
X_abl_full_n[valid_abl_full] = (X_abl_full[valid_abl_full] - abl_x_mean_full) / abl_x_std_full_safe
alpha_ablation_full = np.full(len(test_full), np.nan, dtype=np.float32)
with torch.no_grad():
    pred_valid = ablation_model_full(
        torch.from_numpy(np.nan_to_num(X_abl_full_n[valid_abl_full], nan=0.0)).float().to(DEVICE)
    ).cpu().numpy()
alpha_ablation_full[valid_abl_full] = pred_valid

# Reutilise alpha_model_arr (cellule 12) -- nom corrige, c'est toujours la PROPOSITION du modele,
# pas la decision finale apres filtre de securite.
alpha_mlp_ns_full = alpha_model_arr

strategies_alpha = {
    'EB-priority': alpha_eb_priority_full,
    'Logique floue': alpha_fuzzy_full,
    'MLP simple': alpha_mlp_simple_full,
    f'Ablation ({variant_name_full})': alpha_ablation_full,
    'MLP neurosymbolique (propose, avant filtre)': alpha_mlp_ns_full,
}

print("Strategies disponibles :", list(strategies_alpha.keys()))
print("GNN exclu -- noms reels de la classe et des graphes de test manquants (load_gnn_simple).")

strategies_df = pd.DataFrame({TIME_COL: test_full[TIME_COL], **strategies_alpha})
strategies_df.to_csv(TABLES_DIR / 'results_explicabilite_alpha_toutes_strategies.csv', index=False)
```

```python
# Cellule 15 : explication comparative -- toutes les strategies a un instant donne (pas une seule)
def generer_explication_comparative(i):
    if not valid_full[i] or np.isnan(alpha_ablation_full[i]):
        disponibles = {k: v for k, v in strategies_alpha.items() if not np.isnan(v[i])}
    else:
        disponibles = strategies_alpha

    row = test_full.iloc[i]
    temps, pdem = float(row[TIME_COL]), float(row['hasPower'])

    lignes = [f"--- Comparaison des strategies a l'instant {temps:.0f}s (demande : {pdem/1000:.2f} kW) ---"]
    for nom, valeurs in disponibles.items():
        alpha_val = float(valeurs[i])
        if np.isnan(alpha_val):
            lignes.append(f"  {nom:35s} : donnees indisponibles a cet instant")
            continue
        p_pb_kw, p_eb_kw = alpha_val * pdem / 1000, (1 - alpha_val) * pdem / 1000
        metrics_i = candidate_metrics(alpha_val, pdem, float(row['SOC_EB']), float(row['SOC_PB']), None)
        statut = "respecte les contraintes" if metrics_i['feasible'] else "VIOLE une contrainte physique"
        lignes.append(f"  {nom:35s} : alpha={alpha_val:.3f} | P_PB={p_pb_kw:+.2f}kW, P_EB={p_eb_kw:+.2f}kW | {statut}")

    valeurs_dispo = [v[i] for v in disponibles.values() if not np.isnan(v[i])]
    if len(valeurs_dispo) >= 2:
        ecart = max(valeurs_dispo) - min(valeurs_dispo)
        lignes.append(f"\n  Ecart maximal entre strategies : {ecart:.3f} "
                      f"({'desaccord notable' if ecart > 0.3 else 'strategies globalement alignees'})")

    return "\n".join(lignes)


for i in [10, len(test_full)//2, len(test_full)-10]:
    print(generer_explication_comparative(i))
    print()
```

```python
# Cellule 16 : comparaison des predictions temporelles LSTM vs realite (les 3 sorties, pas seulement Pdem)
def predire_sequence_lstm(model, x_mean, x_std, y_mean, y_std, input_cols, df, window=WINDOW, horizon=HORIZON):
    X = df[input_cols].to_numpy(dtype=np.float32)
    x_std_safe = np.where(np.abs(x_std) < 1e-12, 1.0, x_std)
    Xn = (X - x_mean) / x_std_safe
    n = len(Xn) - window - horizon + 1
    windows = np.stack([Xn[i:i + window] for i in range(n)])
    with torch.no_grad():
        preds_n = model(torch.from_numpy(windows).float().to(DEVICE)).cpu().numpy()
    preds = preds_n * y_std + y_mean
    return preds, n


preds_lstm_seul, n_seul = predire_sequence_lstm(lstm_seul_model, ls_x_mean, ls_x_std, ls_y_mean, ls_y_std, ls_cols, test_full)
preds_lstm_ns, n_ns = predire_sequence_lstm(lstm_ns_model, ln_x_mean, ln_x_std, ln_y_mean, ln_y_std, ln_cols, test_full)

n_commun = min(n_seul, n_ns)
temps_pred = test_full[TIME_COL].iloc[WINDOW + HORIZON - 1 : WINDOW + HORIZON - 1 + n_commun].to_numpy()
pdem_reel = test_full['hasPower'].to_numpy()[WINDOW + HORIZON - 1 : WINDOW + HORIZON - 1 + n_commun]

# Vraies variations de SOC, par soustraction directe -- pas de colonne externe necessaire
soc_eb_full = test_full['SOC_EB'].to_numpy()
soc_pb_full = test_full['SOC_PB'].to_numpy()
anchors = np.arange(WINDOW - 1, len(test_full) - HORIZON)
targets = anchors + HORIZON
delta_soc_eb_reel = soc_eb_full[targets] - soc_eb_full[anchors]
delta_soc_pb_reel = soc_pb_full[targets] - soc_pb_full[anchors]

reels = {'Pdem': pdem_reel, 'delta_SOC_EB': delta_soc_eb_reel[:n_commun], 'delta_SOC_PB': delta_soc_pb_reel[:n_commun]}

output_names_pred = ['Pdem', 'delta_SOC_EB', 'delta_SOC_PB']
fenetre_pred = slice(0, 300)
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
for ax, out_name in zip(axes, output_names_pred):
    idx_out = output_names_pred.index(out_name)
    ax.plot(temps_pred[fenetre_pred], preds_lstm_seul[fenetre_pred, idx_out], label='EMS_LSTM (predit)', alpha=0.8)
    ax.plot(temps_pred[fenetre_pred], preds_lstm_ns[fenetre_pred, idx_out], label='EMS_LSTM_neurosymbolic (predit)', alpha=0.8)
    ax.plot(temps_pred[fenetre_pred], reels[out_name][fenetre_pred], label='Reel', color='black', linestyle='--', alpha=0.6)
    ax.set_title(out_name)
    ax.set_xlabel('Temps (s)')
axes[0].legend(fontsize=8)
axes[0].set_ylabel('Valeur')
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'explicabilite_comparaison_predictions_lstm.png', dpi=300, bbox_inches='tight')
plt.show()

# Note pour le rapport : la baseline Integrated Gradients (cellule 9, torch.zeros_like) correspond,
# dans cet espace normalise, a un etat MOYEN des donnees d'entrainement -- pas a une vitesse/puissance/SOC
# physiquement nulle.
```

```python
# Cellule 19 : regimes physiques + accord/desaccord entre modeles (toutes les strategies disponibles, GNN exclu)

# --- 1. Definition des regimes physiques ---
HIGH_POWER_W = float(globals().get("HIGH_POWER_THRESHOLD_W", 20000.0))
SOC_MARGIN = 0.05

def assigner_regime(row):
    pdem, soc_eb, soc_pb, accel = row['hasPower'], row['SOC_EB'], row['SOC_PB'], row['hasAcceleration']
    if abs(pdem) <= EPS_POWER_W:
        return 'demande_nulle'
    if pdem < -EPS_POWER_W:
        return 'freinage_regeneratif'
    if soc_eb <= SOC_EB_MIN + SOC_MARGIN:
        return 'soc_eb_faible'
    if soc_pb <= SOC_PB_MIN + SOC_MARGIN:
        return 'soc_pb_faible'
    if pdem >= HIGH_POWER_W:
        return 'forte_traction'
    return 'conditions_normales'


regimes = test_full.apply(assigner_regime, axis=1)
print("Distribution des regimes (test) :")
print(regimes.value_counts())
print()
print("Pourcentage :")
print((100 * regimes.value_counts(normalize=True)).round(1))


# --- 2. Alpha de toutes les strategies disponibles (deja calcule en cellule 13) ---
strategies_pour_accord = {k: v for k, v in strategies_alpha.items()}
alpha_matrix = np.stack(list(strategies_pour_accord.values()), axis=1)  # (n, n_strategies)

# accord/desaccord -- ignore les NaN (lignes ou une strategie n'a pas pu produire de valeur)
alpha_std = np.nanstd(alpha_matrix, axis=1)
alpha_mean = np.nanmean(alpha_matrix, axis=1)
n_valides_par_ligne = np.sum(~np.isnan(alpha_matrix), axis=1)

def niveau_accord(std):
    if np.isnan(std):
        return 'indetermine'
    if std < 0.05:
        return 'accord_eleve'
    if std < 0.15:
        return 'accord_moyen'
    return 'desaccord_eleve'

accord_level = np.array([niveau_accord(s) for s in alpha_std])


# --- 3. Assemblage et sauvegarde ---
analyse_regimes_df = pd.DataFrame({
    TIME_COL: test_full[TIME_COL],
    'regime': regimes.to_numpy(),
    'alpha_moyen': alpha_mean,
    'alpha_std': alpha_std,
    'niveau_accord': accord_level,
    'n_strategies_valides': n_valides_par_ligne,
    **{f'alpha_{k}': v for k, v in strategies_pour_accord.items()},
})
analyse_regimes_df.to_csv(TABLES_DIR / 'results_explicabilite_regimes_accord.csv', index=False)

# --- 4. Vue d'ensemble : distribution des regimes ---
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
regimes.value_counts().plot(kind='bar', ax=axes[0])
axes[0].set_title('Distribution des regimes physiques (test)')
axes[0].set_ylabel("Nombre d'observations")
axes[0].tick_params(axis='x', rotation=30)

pd.Series(accord_level).value_counts().plot(kind='bar', ax=axes[1], color=['green', 'orange', 'red'])
axes[1].set_title('Niveau accord entre strategies')
axes[1].set_ylabel("Nombre d'observations")
axes[1].tick_params(axis='x', rotation=15)
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'explicabilite_regimes_et_accord.png', dpi=300, bbox_inches='tight')
plt.show()

# --- 5. Accord moyen PAR regime (table croisee, utile pour la suite) ---
accord_par_regime = (
    pd.DataFrame({'regime': regimes.to_numpy(), 'alpha_std': alpha_std})
    .groupby('regime')['alpha_std']
    .agg(['mean', 'std', 'count'])
    .rename(columns={'mean': 'desaccord_moyen', 'std': 'desaccord_variabilite', 'count': 'n_observations'})
    .sort_values('desaccord_moyen', ascending=False)
)
display(accord_par_regime)
accord_par_regime.to_csv(TABLES_DIR / 'results_explicabilite_desaccord_par_regime.csv')

# --- 6. Les 10 instants de plus fort desaccord ---
top_desaccord = analyse_regimes_df.nlargest(10, 'alpha_std')[[TIME_COL, 'regime', 'alpha_std', 'alpha_moyen'] + [f'alpha_{k}' for k in strategies_pour_accord]]
print("\nLes 10 instants de plus fort desaccord entre strategies :")
display(top_desaccord)
```

```python
# Cellule 21 : tableau de performance par regime, pour les 5 strategies disponibles
alpha_star_full = test_full['alpha_ems_alpha_star'].to_numpy()
valid_demand_full = np.abs(pdem_arr) > EPS_POWER_W

performance_rows = []

for nom_strategie, alpha_vals in strategies_alpha.items():
    valid_strat = ~np.isnan(alpha_vals)
    ev = evaluate_alpha_sequence(test_full, np.nan_to_num(alpha_vals, nan=0.5))

    for regime_nom in regimes.unique():
        mask_regime = (regimes.to_numpy() == regime_nom) & valid_strat & valid_demand_full
        n_obs = int(mask_regime.sum())
        if n_obs == 0:
            continue

        valid_star_regime = mask_regime & ~np.isnan(alpha_star_full)
        mae_vs_star = (
            float(np.mean(np.abs(alpha_vals[valid_star_regime] - alpha_star_full[valid_star_regime])))
            if valid_star_regime.sum() > 0 else np.nan
        )
        rmse_vs_star = (
            float(np.sqrt(np.mean((alpha_vals[valid_star_regime] - alpha_star_full[valid_star_regime]) ** 2)))
            if valid_star_regime.sum() > 0 else np.nan
        )

        performance_rows.append({
            'regime': regime_nom,
            'strategie': nom_strategie,
            'n_observations': n_obs,
            'mae_vs_alpha_star': mae_vs_star,
            'rmse_vs_alpha_star': rmse_vs_star,
            'cout_moyen': float(np.mean(ev['cost'][mask_regime])),
            'faisabilite_pct': float(100 * np.mean(ev['feasible'][mask_regime])),
        })

performance_par_regime_df = pd.DataFrame(performance_rows)

# Regret = cout_moyen - cout_moyen(alpha_star), par regime
alpha_star_ev = evaluate_alpha_sequence(test_full, np.nan_to_num(alpha_star_full, nan=0.5))
cout_star_par_regime = {}
for regime_nom in regimes.unique():
    mask_regime_star = (regimes.to_numpy() == regime_nom) & valid_demand_full & ~np.isnan(alpha_star_full)
    cout_star_par_regime[regime_nom] = float(np.mean(alpha_star_ev['cost'][mask_regime_star])) if mask_regime_star.sum() > 0 else np.nan

performance_par_regime_df['cout_alpha_star_regime'] = performance_par_regime_df['regime'].map(cout_star_par_regime)
performance_par_regime_df['regret_moyen'] = performance_par_regime_df['cout_moyen'] - performance_par_regime_df['cout_alpha_star_regime']

performance_par_regime_df = performance_par_regime_df.sort_values(['regime', 'cout_moyen'])
display(performance_par_regime_df)
performance_par_regime_df.to_csv(TABLES_DIR / 'results_explicabilite_performance_par_regime.csv', index=False)

# Vue synthetique : meilleure strategie (cout le plus bas) par regime
meilleure_par_regime = performance_par_regime_df.loc[performance_par_regime_df.groupby('regime')['cout_moyen'].idxmin()]
print("\nMeilleure strategie par regime (cout physique moyen) :")
display(meilleure_par_regime[['regime', 'strategie', 'n_observations', 'cout_moyen', 'regret_moyen', 'faisabilite_pct']])
```

```python
# Cellule 22 : traitement separe du regime demande_nulle (Pdem quasi nul -- regret peu pertinent ici)
mask_demande_nulle = regimes.to_numpy() == 'demande_nulle'

demande_nulle_rows = []
for nom_strategie, alpha_vals in strategies_alpha.items():
    valid_strat = ~np.isnan(alpha_vals) & mask_demande_nulle
    n_obs = int(valid_strat.sum())
    if n_obs == 0:
        continue

    # Faisabilite -- toujours calculable et pertinente, meme a Pdem quasi nul
    feasible_count = 0
    for i in np.flatnonzero(valid_strat):
        m = candidate_metrics(float(alpha_vals[i]), float(pdem_arr[i]), float(soc_eb_arr[i]), float(soc_pb_arr[i]), None)
        feasible_count += int(m['feasible'])

    demande_nulle_rows.append({
        'strategie': nom_strategie,
        'n_observations': n_obs,
        'alpha_moyen': float(np.mean(alpha_vals[valid_strat])),
        'alpha_std': float(np.std(alpha_vals[valid_strat])),  # stabilite -- une grande dispersion ici est suspecte
        'faisabilite_pct': 100.0 * feasible_count / n_obs,
    })

demande_nulle_df = pd.DataFrame(demande_nulle_rows).sort_values('faisabilite_pct', ascending=False)
display(demande_nulle_df)
demande_nulle_df.to_csv(TABLES_DIR / 'results_explicabilite_regime_demande_nulle.csv', index=False)
```

```python
# Cellule 23 : matrice de decision normalisee, deux profils (securite / performance)
# Metriques disponibles uniquement : regret_moyen, mae_vs_alpha_star, taux_violation, cout_moyen
# P_unserved et instabilite temporelle NE SONT PAS calcules -- poids renormalises sur ce qui existe

df_matrice = performance_par_regime_df.copy()
df_matrice['taux_violation'] = 100.0 - df_matrice['faisabilite_pct']

METRIQUES_A_MINIMISER = ['regret_moyen', 'mae_vs_alpha_star', 'taux_violation', 'cout_moyen']

def normaliser_par_regime(df, col):
    return df.groupby('regime')[col].transform(lambda x: (x - x.min()) / (x.max() - x.min() + 1e-12))

for col in METRIQUES_A_MINIMISER:
    df_matrice[f'z_{col}'] = normaliser_par_regime(df_matrice, col)

# Poids originaux du document, renormalises sur les seules metriques disponibles
POIDS_SECURITE = {'z_taux_violation': 0.35, 'z_regret_moyen': 0.20, 'z_mae_vs_alpha_star': 0.10}
total_secu = sum(POIDS_SECURITE.values())
POIDS_SECURITE = {k: v / total_secu for k, v in POIDS_SECURITE.items()}

POIDS_PERFORMANCE = {'z_regret_moyen': 0.35, 'z_mae_vs_alpha_star': 0.25, 'z_cout_moyen': 0.20, 'z_taux_violation': 0.15}
total_perf = sum(POIDS_PERFORMANCE.values())
POIDS_PERFORMANCE = {k: v / total_perf for k, v in POIDS_PERFORMANCE.items()}

print("Poids securite (renormalises) :", {k: round(v, 3) for k, v in POIDS_SECURITE.items()})
print("Poids performance (renormalises) :", {k: round(v, 3) for k, v in POIDS_PERFORMANCE.items()})

df_matrice['score_securite'] = sum(df_matrice[k] * w for k, w in POIDS_SECURITE.items())
df_matrice['score_performance'] = sum(df_matrice[k] * w for k, w in POIDS_PERFORMANCE.items())

print("\n=== Meilleure strategie par regime, profil SECURITE ===")
meilleur_secu = df_matrice.loc[df_matrice.groupby('regime')['score_securite'].idxmin()]
display(meilleur_secu[['regime', 'strategie', 'score_securite', 'taux_violation', 'regret_moyen']])

print("\n=== Meilleure strategie par regime, profil PERFORMANCE ===")
meilleur_perf = df_matrice.loc[df_matrice.groupby('regime')['score_performance'].idxmin()]
display(meilleur_perf[['regime', 'strategie', 'score_performance', 'cout_moyen', 'regret_moyen']])

print("\n=== Les deux profils sont-ils d'accord ? ===")
accord_profils = meilleur_secu[['regime', 'strategie']].merge(
    meilleur_perf[['regime', 'strategie']], on='regime', suffixes=('_securite', '_performance')
)
accord_profils['profils_d_accord'] = accord_profils['strategie_securite'] == accord_profils['strategie_performance']
display(accord_profils)

df_matrice.to_csv(TABLES_DIR / 'results_explicabilite_matrice_decision.csv', index=False)
accord_profils.to_csv(TABLES_DIR / 'results_explicabilite_matrice_accord_profils.csv', index=False)
```

```python
# Cellule 24 : synthese textuelle finale -- recommandation par regime, combinant les deux profils

REGIME_LABELS = {
    'conditions_normales': 'conditions normales de conduite',
    'forte_traction': 'forte traction (accélération soutenue)',
    'freinage_regeneratif': 'freinage régénératif',
    'soc_eb_faible': 'SOC de la batterie d\'énergie faible',
    'demande_nulle': 'demande de puissance quasi nulle',
}


def generer_synthese_regime(regime_nom):
    ligne_secu = meilleur_secu[meilleur_secu['regime'] == regime_nom].iloc[0]
    ligne_perf = meilleur_perf[meilleur_perf['regime'] == regime_nom].iloc[0]
    accord = accord_profils[accord_profils['regime'] == regime_nom].iloc[0]['profils_d_accord']

    n_obs = int(performance_par_regime_df[performance_par_regime_df['regime'] == regime_nom]['n_observations'].iloc[0])
    pct_test = 100 * n_obs / len(test_full)
    label = REGIME_LABELS.get(regime_nom, regime_nom)

    texte = f"--- {label.upper()} ({n_obs} observations, {pct_test:.1f}% du test) ---\n"

    if accord:
        texte += (f"Les profils securite et performance s'accordent : {ligne_perf['strategie']} est recommande, "
                   f"avec un regret moyen de {ligne_perf['regret_moyen']:.4f} et un taux de violation de "
                   f"{ligne_secu['taux_violation']:.1f}%.")
    else:
        texte += (
            f"Les profils divergent. Le profil performance recommande {ligne_perf['strategie']} "
            f"(regret moyen {ligne_perf['regret_moyen']:.4f}, le plus proche de l'optimum physique), "
            f"mais le profil securite recommande plutot {ligne_secu['strategie']} "
            f"(taux de violation {ligne_secu['taux_violation']:.1f}%, contre un taux nettement plus eleve "
            f"pour les modeles neuronaux dans ce regime). "
            f"Ceci illustre un compromis reel entre proximite a l'optimum theorique et fiabilite physique : "
            f"le modele le plus performant n'est pas necessairement le plus sur a deployer sans filtre de securite."
        )

    return texte


print("=" * 90)
print("SYNTHESE : RECOMMANDATION DE STRATEGIE PAR REGIME PHYSIQUE")
print("=" * 90)

synthese_rows = []
for regime_nom in performance_par_regime_df['regime'].unique():
    texte_regime = generer_synthese_regime(regime_nom)
    print()
    print(texte_regime)
    synthese_rows.append({'regime': regime_nom, 'synthese': texte_regime})

synthese_df = pd.DataFrame(synthese_rows)
synthese_df.to_csv(TABLES_DIR / 'results_explicabilite_synthese_par_regime.csv', index=False)

print("\n" + "=" * 90)
print("CONCLUSION GENERALE")
print("=" * 90)
n_divergences = (~accord_profils['profils_d_accord']).sum()
print(f"\nSur {len(accord_profils)} regimes analyses (hors demande_nulle, traite separement), "
      f"{n_divergences} presente(nt) une divergence entre performance et securite : "
      f"{', '.join(accord_profils[~accord_profils['profils_d_accord']]['regime'].tolist())}.")
print("Ce resultat justifie empiriquement la necessite d'un filtre de securite physique en aval "
      "des modeles neuronaux, plutot que leur deploiement direct, meme lorsque ces modeles "
      "approchent mieux l'optimum theorique que les regles simples comme EB-priority.")
```
